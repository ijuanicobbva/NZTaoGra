## Process

```scala
import Constants._
import org.apache.spark.sql.functions.{abs, broadcast, coalesce, col, concat, current_timestamp, lit, lpad, trim, upper, when}
import org.apache.spark.sql.DataFrame
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.types.{StringType, DateType, DecimalType}
import scala.util.Random


class Process(odate: String) {
  
  def fixedRentActive(rentaFijaOpeBis: DataFrame, catalogA050: DataFrame, catalogA219: DataFrame,
                      tsiContraparte: DataFrame): DataFrame = {
    try {
      rentaFijaOpeBis
        .join(broadcast(catalogA050), catalogA050.col(gf_catalog_val_id) === col(source_application_id), left)
        .join(broadcast(catalogA219), catalogA219.col(gf_catalog_val_id) === col(securities_dealing_desk_name), left)
        .join(tsiContraparte, Seq(counterparty_id), left)
        .select(
          coalesce(col(application_desc), lit(BGM_BIS)).as(gf_origin_application_id),
          concat(lit(MX0074), when(col(seller_branch_id).isNull, fourZero).
            otherwise(lpad(trim(col(seller_branch_id)), four, zeroString))).as(g_branch_id),
          when(col(gf_secur_pos_registration_date) === lit(emptyString), lit(nullString)).
            otherwise(col(gf_secur_pos_registration_date)).as(gf_secur_pos_registration_date),
          col(issuing_expiry_date).as(gf_current_expiration_date),
          when(col(origin_currency_type) === UDI, lit(MXV))
            .otherwise(col(origin_currency_type)).as(g_currency_id),
          col(isin_id).as(g_security_id),
          col(g_security_position_id),
          lit(MX0074).as(g_entity_id),
          when(col(market_operation_type).isin(prods2:_*), concat(lit(MX0074), lit(fourZero), col(var_coverage_secty_mark_type), col(isin_id))).
            otherwise(concat(lit(MX0074),
              when(col(seller_branch_id).isNull, fourZero).otherwise(lpad(col(seller_branch_id), four, zeroString)),
              col(contract_id))).as(gf_local_contract_number_id),
          col(market_operation_type).as(gf_financial_product_id),
          lit("1611101010").as(g_global_product_id),
          lit(RDR).as(gf_ctpty_source_system_id),
          lit(RDR).as(gf_srce_system_counterparty_id),
          when(col(counterparty_id) === lit(emptyString), lit(nullString)).
            otherwise(col(counterparty_id)).as(g_counterparty_id),
          when(col(contract_affiliation_id).isNull , lit(nineZero))
            .when(col(contract_affiliation_id) === emptyString, lit(nineZero))
            .otherwise(lpad(col(contract_affiliation_id), nine, zeroString)).as(g_customer_id),
          abs(col(total_securities_nominal_amount)).as(gf_nominal_amount),
          col(current_commitment_amount).as(gf_secur_pos_avg_trd_pc_amount),
          col(return_interest_rate_per).as(gf_irr_per),
          col(current_rate_coupon_per).as(gf_current_interest_rate_per),
          when(col(market_operation_type).isNull, lit(X))
            .when(col(market_operation_type) === emptyString, lit(X))
            .when(col(market_operation_type).isin(CET, CVT, LMG, PVR, CDP, CNG), lit(C))
            .when(col(market_operation_type).isin(SPR, PRB, GPR), lit(V))
            .otherwise(col(market_operation_type)).as(g_trade_position_type),
          col(portfolio_desc).as(gf_management_portfolio_id),
          lit(nullString).as(gf_option_strike_amount),
          lit(nullString).as(g_exercise_option_type),
          lit(nullString).as(g_call_put_option_type),
          lit(nullString).as(gf_undly_prd_typo_desc),
          col(isin_id).as(g_undly_asset_isin_id),
          lit(nullString).as(gf_colat_typo_desc),
          lit(nullString).as(gf_collateral_isin_desc),
          lit(nullString).as(gf_colat_nmnl_amt_desc),
          col(initial_amount).as(gf_settl_amount),
          lit(BILT).as(g_mic_id),
          when(col(market_operation1_type) === E, lit(N))
            .when(col(market_operation1_type) === I, lit(Y))
            .otherwise(col(market_operation1_type)).as(g_internal_operation_ind_type),
          lit(Y).as(gf_trd_cpt_cnsptn_pr_ind_type),
          lit(nullString).as(gf_deriv_setl_mkt_ind_type),
          lit(nullString).as(gf_recouponing_date),
          col(current_rate_coupon_per).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per),
          col(issuing_expiry_date).as(gf_original_expiry_date),
          col(coupon_rate_name).as(g_reference_index_id),
          col(return_interest_rate_per).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
    
  def fixedRentPasive(rentFijPasPosBis001: DataFrame, catalogA050: DataFrame, catalogA219: DataFrame, tsiContraparte: DataFrame): DataFrame = {
    try {
      rentFijPasPosBis001.join(broadcast(catalogA050), catalogA050.col(gf_catalog_val_id) === col(source_application_id), left)
        .join(broadcast(catalogA219), catalogA219.col(gf_catalog_val_id) === col(securities_dealing_desk_name), left)
        .join(tsiContraparte, Seq(counterparty_id), left)
        .select(
          coalesce(col(application_desc), lit(BGM_BIS)).as(gf_origin_application_id),
          concat(lit(MX0074), when(col(seller_branch_id).isNull, fourZero)
            .otherwise(lpad(trim(col(seller_branch_id)), four, zeroString))).as(g_branch_id),
          col(gf_secur_pos_registration_date),
          col(issuing_expiry_date).as(gf_current_expiration_date),
          when(col(origin_currency_type) === UDI, lit(MXV))
            .otherwise(col(origin_currency_type)).as(g_currency_id),
          col(isin_id).as(g_security_id),
          col(g_security_position_id),
          lit(MX0074).as(g_entity_id),
          concat(lit(MX0074), when(col(seller_branch_id).isNull, fourZero)
            .otherwise(lpad(col(seller_branch_id), four, zeroString)), col(contract_id)).as(gf_local_contract_number_id),
          col(market_operation_type).as(gf_financial_product_id),
          lit("1611101010").as(g_global_product_id),
          lit(RDR).as(gf_ctpty_source_system_id),
          lit(RDR).as(gf_srce_system_counterparty_id),
          when(col(counterparty_id) === lit(emptyString), lit(nullString)).
            otherwise(col(counterparty_id)).as(g_counterparty_id),
          when(col(contract_affiliation_id).isNull , lit(nineZero))
            .when(col(contract_affiliation_id) === emptyString, lit(nineZero))
            .otherwise(lpad(col(contract_affiliation_id), nine, zeroString)).as(g_customer_id),
          abs(col(total_securities_nominal_amount)).as(gf_nominal_amount),
          col(current_commitment_amount).as(gf_secur_pos_avg_trd_pc_amount),
          lit(nullString).as(gf_irr_per),
          col(current_rate_coupon_per).as(gf_current_interest_rate_per),
          when(col(market_operation_type).isNull, lit(X))
            .when(col(market_operation_type) === emptyString, lit(X))
            .when(col(market_operation_type) === VDO, lit(V))
            .otherwise(col(market_operation_type)).as(g_trade_position_type),
          col(portfolio_desc).as(gf_management_portfolio_id),
          lit(nullString).as(gf_option_strike_amount),
          lit(nullString).as(g_exercise_option_type),
          lit(nullString).as(g_call_put_option_type),
          lit(nullString).as(gf_undly_prd_typo_desc),
          col(isin_id).as(g_undly_asset_isin_id),
          lit(nullString).as(gf_colat_typo_desc),
          lit(nullString).as(gf_collateral_isin_desc),
          lit(nullString).as(gf_colat_nmnl_amt_desc),
          col(initial_amount).as(gf_settl_amount),
          lit(BILT).as(g_mic_id),
          when(col(market_operation1_type) === E, lit(N))
            .when(col(market_operation1_type) === I, lit(Y))
            .otherwise(col(market_operation1_type)).as(g_internal_operation_ind_type),
          lit(E).as(gf_trd_cpt_cnsptn_pr_ind_type),
          lit(nullString).as(gf_deriv_setl_mkt_ind_type),
          lit(nullString).as(gf_recouponing_date),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per),
          lit(nullString).cast(DateType).as(gf_original_expiry_date),
          lit(nullString).cast(StringType).as(g_reference_index_id),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
    
  def fixedRent(fixedRentActive: DataFrame, fixedRentPasive: DataFrame): DataFrame = {
    try {
      fixedRentActive.union(fixedRentPasive)
        .select(
          lit(MX).as(g_entific_id),
          col(gf_origin_application_id),
          lit(odate).cast(DateType).as(gf_cutoff_date),
          lit(MX0074).as(g_entity_id),
          col(g_branch_id),
          col(gf_secur_pos_registration_date).cast(DateType),
          col(gf_current_expiration_date),
          lit(nullString).cast(DateType).as(gf_secur_pos_cancel_date),
          lit(nullString).cast(DateType).as(gf_secur_pos_last_change_date),
          lit(nullString).cast(StringType).as(gf_secur_posn_status_type),
          lit(nullString).cast(StringType).as(gf_secur_pos_ast_liab_ind_type),
          col(g_currency_id),
          col(g_security_id),
          col(g_security_position_id), lit(nullString).cast(StringType).as(gf_security_posn_id), col(gf_local_contract_number_id),
          lit(nullString).cast(StringType).as(gf_secur_asset_type), col(gf_financial_product_id),
          lit(nullString).cast(StringType).as(g_holding_product_id), col(g_global_product_id), col(gf_ctpty_source_system_id),
          col(gf_srce_system_counterparty_id), col(g_counterparty_id), col(g_customer_id),
          col(gf_nominal_amount).cast(DecimalType(twentyeight, eight)), lit(nullString).cast(DecimalType(twentysix, six)).as(gf_lc_nominal_amount),
          col(gf_secur_pos_avg_trd_pc_amount).cast(DecimalType(twentysix, six)), lit(nullString).cast(DecimalType(twelve, nine)).as(gf_secur_posn_avg_irr_per),
          col(gf_irr_per).cast(DecimalType(twelve, nine)), lit(nullString).cast(DecimalType(twelve, nine)).as(gf_tedr_per),
          col(gf_current_interest_rate_per).cast(DecimalType(twelve, nine)), lit(nullString).cast(DecimalType(twelve, nine)).as(gf_nominal_interest_rate_per),
          lit(nullString).cast(DecimalType(twentysix, six)).as(gf_gl_secur_pc_amount),
          lit(nullString).cast(DecimalType(seventeen, zero)).as(gf_securities_number), lit(nullString).cast(StringType).as(g_bought_sold_position_type),
          col(g_trade_position_type), col(gf_management_portfolio_id),
          lit(nullString).cast(DecimalType(twelve, four)).as(gf_contract_multiplier_number), col(gf_option_strike_amount).cast(DecimalType(twentysix, six)),
          col(g_exercise_option_type).cast(StringType), col(g_call_put_option_type).cast(StringType), col(gf_undly_prd_typo_desc).cast(StringType),
          col(g_undly_asset_isin_id), lit(nullString).cast(DateType).as(gf_underlying_maturity_date),
          col(gf_colat_typo_desc).cast(StringType), col(gf_collateral_isin_desc).cast(StringType), col(gf_colat_nmnl_amt_desc).cast(StringType),
          lit(nullString).cast(StringType).as(g_colat_nmnl_amt_curncy_desc), col(gf_settl_amount).cast(DecimalType(twentysix, six)),
          lit(nullString).cast(StringType).as(g_settl_curncy_id), col(g_mic_id), col(g_internal_operation_ind_type),
          col(gf_trd_cpt_cnsptn_pr_ind_type), col(gf_deriv_setl_mkt_ind_type).cast(StringType), col(gf_recouponing_date).cast(DateType),
          col(g_triparty_trade_ind_type).cast(StringType).as(g_triparty_trade_ind_type), lit(nullString).cast(StringType).as(g_clearing_house_type),
          lit(nullString).cast(StringType).as(g_gl_classification_type), lit(nullString).cast(StringType).as(g_market_type),
          col(gf_initial_interest_rate_per).as(gf_initial_interest_rate_per),
          lit(nullString).cast(StringType).as(gf_float_rt_index_id), col(g_reference_index_id).as(g_reference_index_id),
          col(gf_ref_index_interest_rate_per).as(gf_ref_index_interest_rate_per), col(gf_original_expiry_date).as(gf_original_expiry_date),
          current_timestamp.as(gf_audit_date), lit(odate.replace("-", emptyString)).as(gf_odate_date_id),
          concat(lit(odate.replace("-", emptyString)), lit(Random.alphanumeric take 11 mkString)).as(gf_run_id)
        ).filter(!col(gf_financial_product_id).isin(VDO, PVR))
        .orderBy(col(gf_local_contract_number_id), col(gf_secur_pos_registration_date).asc_nulls_last)
        .dropDuplicates(gf_local_contract_number_id)
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
    
  def securAccountingPosition(fixedRent: DataFrame, subsidiaries: DataFrame, flagSubsidiaries: String = "F"): DataFrame = {
    try {
      if (flagSubsidiaries == "T") {
        clientHomologateBank(fixedRent.union(subsidiaries),List(g_customer_id))
      } else {
        clientHomologateBank(fixedRent,List(g_customer_id))
      }
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
    
}
```
