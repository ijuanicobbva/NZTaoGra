```scala
import Constants._
import org.apache.spark.sql.functions.{broadcast, coalesce, col, lit, trim, when, min}
import org.apache.spark.sql.{DataFrame, SparkSession}
import org.apache.commons.lang3.exception.ExceptionUtils
class InputReader(spark: SparkSession, hdate: String) {
  val sparkI = spark
  import sparkI.implicits._
  def readCatalogApp(path: String): DataFrame = {
    try {
      spark.read.parquet(path).filter(col(g_catalog_id) === A050)
        .select(
          col(gf_catalog_val_id).as(id),
          col(gf_catlg_field_value_es_desc).as(value)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def catalogFlowsType(): DataFrame = {
    try {
      val catalogFlowsType = Seq(
        ("PI","Adicional","ADD"),
        ("CA","Cancelacion","CA"),
        ("CL","Sin Descripción","ND"),
        ("CP","Adicional","ADD"),
        ("EX","Prima por Ejercicio","EXR"),
        ("FE","Broker fee","FEE"),
        ("FI","Sin Descripción","ND"),
        ("FL","Interes","INT"),
        ("FW","Prima por Market","FLW"),
        ("IN","Intercambio Nocionales","CAP"),
        ("LI","Sin Descripción","ND"),
        ("MO","Sin Descripción","ND"),
        ("PR","Prima Trade","PRE"),
        ("FS","Flujo superpuesto","FSP"),
        ("FN","Non Delivery Forward","FL_NDF")
      ).toDF("LOCAL","DESCRIPCION","MUREX")
      catalogFlowsType
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readTsiOperacion(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          trim(col(operation_sire_assigned_id)).as(operation_sire_assigned_id),
          col(oper_origin_app_assigned_id).as(operation_deal_front_star_id),
          trim(col(financial_product_id)).as(financial_product_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def flagSwap(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          col(operation_sire_assigned_id),
          col(fin_derv_flow_phases_number)
        )
        .groupBy(operation_sire_assigned_id)
        .agg(
          min(col(fin_derv_flow_phases_number)).as(validate)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readLiquidityFlowsOverSwap(path:String, flagSwap: DataFrame): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .join(flagSwap, Seq(operation_sire_assigned_id), left)
        .select(
          trim(col(operation_sire_assigned_id)).as(operation_sire_assigned_id),
          col(murex_operation_id).as(operation_deal_front_star_id),
          coalesce((col(swap_legs_id) + 1), lit(oneString)).as(legs_id),
          when(col(fin_derv_flow_phases_number).isNull, lit(one))
            .when(col(validate) === zero, col(fin_derv_flow_phases_number) + one)
            .otherwise(col(fin_derv_flow_phases_number)).as(phases_number),
          trim(col(phase_flow_leg_sequential_id)).as(phase_flow_leg_sequential_id),
          trim(col(fin_derv_oper_curr_flow_type)).as(fin_derv_oper_curr_flow_type),
          col(fin_derv_int_notl_rgtry_date),
          col(period_end_date),
          col(fix_date),
          col(fixed_cost_amount),
          col(payment_date),
          col(purchase_price_amount),
          col(diffc_exch_curncy_oper_amount),
          col(fixed_var_flow_leg_mark_type),
          col(period_end_date).as(underlying_expiry_date),
          col(fin_derv_flow_start_date),
          col(interest_rate_type).as(settlement_flow_type),
          when(col(oper_flow_pymt_rcpt_mark_type) === P, PAYER)
            .when(col(oper_flow_pymt_rcpt_mark_type) === R, RECEIVER).as(oper_flow_pymt_rcpt_mark_type),
          col(fndr_flow_pymt_premium_amount),
          col(fin_derv_notional_int_amount),
          col(fin_derv_addl_flow_curncy_id),
          col(notional_currency_id),
          col(fndr_notl_exch_curncy_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def flagOptions(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          col(operation_sire_assigned_id),
          col(fin_derv_flow_phases_number)
        )
        .groupBy(operation_sire_assigned_id)
        .agg(
          min(col(fin_derv_flow_phases_number)).as(validate)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readLiqOverRateOptions(path:String, flagOptions: DataFrame): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .join(flagOptions, Seq(operation_sire_assigned_id), left)
        .select(
          col(operation_sire_assigned_id),
          col(oper_origin_app_assigned_id).as(operation_deal_front_star_id),
          (col(fin_derv_flow_phases_number) + 1).as(legs_id),
          when(col(fin_derv_flow_phases_number).isNull, lit(1))
            .when(col(validate) === zero, col(fin_derv_flow_phases_number) + one)
            .otherwise(col(fin_derv_flow_phases_number)).as(phases_number),
          col(phase_flow_leg_sequential_id),
          trim(col(fin_derv_oper_curr_flow_type)).as(fin_derv_oper_curr_flow_type),
          lit(nullString).as(fin_derv_int_notl_rgtry_date),
          lit(nullString).as(period_end_date),
          col(fix_date),
          col(fixed_cost_amount),
          col(payment_date),
          col(securities_nominal_amount),
          col(current_spread_amount),
          col(fixed_var_flow_leg_mark_type),
          col(underlying_expiry_date),
          col(fin_derv_flow_start_date),
          col(settlement_flow_type),
          when(col(oper_flow_pymt_rcpt_mark_type) === P, PAYER)
            .when(col(oper_flow_pymt_rcpt_mark_type) === R, RECEIVER).as(oper_flow_pymt_rcpt_mark_type),
          lit(nullString).as(fndr_flow_pymt_premium_amount),
          lit(nullString).as(fin_derv_notional_int_amount),
          lit(nullString).as(fin_derv_addl_flow_curncy_id),
          lit(nullString).as(notional_currency_id),
          lit(nullString).as(fndr_notl_exch_curncy_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readMoneyMarketDeposits(path:String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          col(operation_sire_assigned_id),
          col(oper_origin_app_assigned_id).as(operation_deal_front_star_id),
          lit(oneString).as(legs_id),
          lit(oneString).as(phases_number),
          col(phase_flow_leg_sequential_id),
          col(fin_derv_oper_curr_flow_type),
          col(fin_derv_int_notl_rgtry_date),
          lit(nullString).as(period_end_date),
          lit(emptyString).as(fix_date),
          col(fixed_cost_amount),
          col(payment_date),
          col(securities_nominal_amount),
          col(current_spread_amount),
          col(fixed_var_flow_leg_mark_type),
          col(underlying_expiry_date),
          col(fin_derv_flow_start_date),
          col(settlement_flow_type),
          when(col(oper_flow_pymt_rcpt_mark_type) === P, PAYER)
            .when(col(oper_flow_pymt_rcpt_mark_type) === R, RECEIVER).as(oper_flow_pymt_rcpt_mark_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readDerivadosMmooBis(path: String, catalogApp: DataFrame): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate).filter(col(product_type).isin(flowProd:_*))
        .select(
          trim(col(source_application_id)).as(source_application_id),
          col(operation_deal_front_star_id),
          col(operation_deal_back_star_id)
        )
        .join(broadcast(catalogApp), col(source_application_id) === col(id), left)
        .select(
          when(coalesce(col(value), col(source_application_id)) === emptyString, BGM_BIS)
            .otherwise(coalesce(col(value), col(source_application_id))).as(source_application_id),
          col(operation_deal_front_star_id),
          col(operation_deal_back_star_id)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readFinDervOtcBis(path: String, catalogApp: DataFrame): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate).filter(col(subproduct_type).isin(flowProd:_*))
        .select(
          trim(col(source_application_id)).as(source_application_id),
          col(operation_deal_front_star_id),
          col(operation_deal_back_star_id)
        )
        .join(broadcast(catalogApp), trim(col(source_application_id)) === col(id), left)
        .select(
          when(coalesce(col(value), col(source_application_id)) === emptyString, BGM_BIS)
            .otherwise(coalesce(col(value), col(source_application_id))).as(source_application_id),
          col(operation_deal_front_star_id),
          col(operation_deal_back_star_id)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def readInternalOperMurex(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate).filter(col(transaction_product_type) === LoansDeposits)
        .select(
          lit(MUREX).as(source_application_id),
          col(front_system_number_id).as(operation_deal_front_star_id),
          lit(emptyString).as(operation_deal_back_star_id)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
}
```

## Process

```scala
import Constants._
import org.apache.spark.sql.functions.{coalesce, col, concat, current_timestamp, lit, trim, when, regexp_replace}
import org.apache.spark.sql.DataFrame
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.types.{DateType, DecimalType, StringType, IntegerType}
import scala.util.Random
class Process(odate: String) {
   def flowsSwpOpt(liquidityFlowsOverSwap: DataFrame, liqOverRateOptions: DataFrame, tsiOperacion: DataFrame): DataFrame = {
    try {
      liquidityFlowsOverSwap.union(liqOverRateOptions)
        .join(tsiOperacion, Seq(operation_sire_assigned_id, operation_deal_front_star_id))
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def flowsDeposOperacion(moneyMarketDeposits: DataFrame, tsiOperacion: DataFrame): DataFrame = {
    try {
      moneyMarketDeposits.join(tsiOperacion, Seq(operation_sire_assigned_id, operation_deal_front_star_id))
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def flowsMmo(derivadosMmooBis: DataFrame, flowsSwpOpt: DataFrame, catalogApp: DataFrame, catalogFlowsType: DataFrame): DataFrame = {
    try {
      derivadosMmooBis.join(flowsSwpOpt, Seq(operation_deal_front_star_id))
        .join(catalogApp, col(source_application_id) === col(id), left)
        .join(catalogFlowsType, col(fin_derv_oper_curr_flow_type) === col(LOCAL), left)
        .select(
          col(source_application_id).as(gf_origin_application_id),
          col(source_application_id).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id),
          col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          concat(lit(MX0074),
            coalesce(col(value), col(source_application_id)),
            when(trim(col(operation_deal_front_star_id)) === "0", trim(col(operation_deal_back_star_id)))
              .otherwise(col(operation_deal_front_star_id))).as(gf_local_contract_number_id),
          col(fixed_var_flow_leg_mark_type).as(g_interest_rate_type),
          col(underlying_expiry_date).as(gf_calc_p_adj_end_date),
          col(fin_derv_flow_start_date).as(gf_calc_p_adj_start_date),
          col(settlement_flow_type).as(gf_float_rt_index_id),
          col(oper_flow_pymt_rcpt_mark_type).as(gf_cshflw_main_ctpty_role_type),
          col(legs_id).as(gf_trd_leg_id),
          col(phases_number).as(gf_trd_phase_id),
          col(phase_flow_leg_sequential_id).as(gf_trd_cshflw_bkwd_ofst_number),
          col(MUREX).as(gf_cshflw_typo_desc),
          when(col(fixed_var_flow_leg_mark_type) === F, lit(emptyString)).otherwise(col(fix_date)).as(gf_fixing_date),
          when(col(fixed_cost_amount) === zero, nullString).otherwise(col(fixed_cost_amount)).as(gf_ref_index_interest_rate_per),
          coalesce(regexp_replace(col(payment_date), date_1900_01_01,emptyString), lit(emptyString)).as(gf_adjusted_pmt_date),
          col(purchase_price_amount).as(gf_ctrct_interest_rate_per),
          col(diffc_exch_curncy_oper_amount).as(gf_int_rate_spread_margin_per),
          when(col(fin_derv_oper_curr_flow_type).isNull || col(fin_derv_oper_curr_flow_type).isin(productsFlowsMmo:_*), col(fndr_flow_pymt_premium_amount))
            .when(col(fin_derv_oper_curr_flow_type) === lit(IN), col(fin_derv_notional_int_amount)).as(gf_cshflw_amount),
          when(col(fin_derv_oper_curr_flow_type).isNull || col(fin_derv_oper_curr_flow_type).isin(productsFlowsMmo2:_*), col(fin_derv_addl_flow_curncy_id))
            .when(col(fin_derv_oper_curr_flow_type) === lit(CL), col(notional_currency_id))
             .when(col(fin_derv_oper_curr_flow_type) === lit(IN), col(fndr_notl_exch_curncy_type))
              .otherwise(fin_derv_oper_curr_flow_type).as(g_cshflw_curncy_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def flowsOtc(finDervOtcBis: DataFrame, flowsSwpOpt: DataFrame, catalogApp: DataFrame, catalogFlowsType: DataFrame): DataFrame = {
    try {
      finDervOtcBis.join(flowsSwpOpt, Seq(operation_deal_front_star_id))
        .join(catalogApp, col(source_application_id) === col(id), left)
        .join(catalogFlowsType, col(fin_derv_oper_curr_flow_type) === col(LOCAL), left)
        .select(
          col(source_application_id).as(gf_origin_application_id),
          col(source_application_id).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id),
          col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          concat(lit(MX0074),
            coalesce(col(value), col(source_application_id)),
            trim(col(operation_deal_front_star_id))).as(gf_local_contract_number_id),
          col(fixed_var_flow_leg_mark_type).as(g_interest_rate_type),
          col(underlying_expiry_date).as(gf_calc_p_adj_end_date),
          col(fin_derv_flow_start_date).as(gf_calc_p_adj_start_date),
          col(settlement_flow_type).as(gf_float_rt_index_id),
          col(oper_flow_pymt_rcpt_mark_type).as(gf_cshflw_main_ctpty_role_type),
          col(legs_id).as(gf_trd_leg_id),
          col(phases_number).as(gf_trd_phase_id),
          col(phase_flow_leg_sequential_id).as(gf_trd_cshflw_bkwd_ofst_number),
          col(MUREX).as(gf_cshflw_typo_desc),
          when(col(fixed_var_flow_leg_mark_type) === F, lit(emptyString)).otherwise(col(fix_date)).as(gf_fixing_date),
          when(col(fixed_cost_amount) === zero, nullString).otherwise(col(fixed_cost_amount)).as(gf_ref_index_interest_rate_per),
          regexp_replace(when(col(fin_derv_oper_curr_flow_type).isin(IN,CP), col(fin_derv_int_notl_rgtry_date))
            .when(col(fin_derv_oper_curr_flow_type).isin(CL,PI), col(period_end_date))
            .otherwise(col(payment_date)), date_1900_01_01, emptyString).as(gf_adjusted_pmt_date),
          col(purchase_price_amount).as(gf_ctrct_interest_rate_per),
          col(diffc_exch_curncy_oper_amount).as(gf_int_rate_spread_margin_per),
          when(col(fin_derv_oper_curr_flow_type).isNull || col(fin_derv_oper_curr_flow_type).isin(productsFlowsMmo:_*), col(fndr_flow_pymt_premium_amount))
            .when(col(fin_derv_oper_curr_flow_type) === lit(IN), col(fin_derv_notional_int_amount))
             .otherwise(fin_derv_oper_curr_flow_type).as(gf_cshflw_amount),
          when(col(fin_derv_oper_curr_flow_type).isNull || col(fin_derv_oper_curr_flow_type).isin(productsFlowsMmo2:_*), col(fin_derv_addl_flow_curncy_id))
            .when(col(fin_derv_oper_curr_flow_type) === lit(CL), col(notional_currency_id))
             .when(col(fin_derv_oper_curr_flow_type) === lit(IN), col(fndr_notl_exch_curncy_type))
              .otherwise(fin_derv_oper_curr_flow_type).as(g_cshflw_curncy_id)
        )
    } catch {
      case e:Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def flowsDeposits(internalOperMurex: DataFrame, flowsDeposOperacion: DataFrame, catalogFlowsType: DataFrame): DataFrame = {
    try {
      internalOperMurex.join(flowsDeposOperacion, Seq(operation_deal_front_star_id))
        .join(catalogFlowsType, col(fin_derv_oper_curr_flow_type) === col(LOCAL), left)
        .select(
          col(source_application_id).as(gf_origin_application_id),
          col(source_application_id).as(gf_front_sys_name),
          col(operation_deal_front_star_id).cast(StringType).as(gf_front_sys_trd_id),
          col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          concat(lit(MX0074 + MUREX), col(operation_deal_front_star_id)).as(gf_local_contract_number_id),
          col(fixed_var_flow_leg_mark_type).as(g_interest_rate_type),
          col(underlying_expiry_date).as(gf_calc_p_adj_end_date),
          col(fin_derv_flow_start_date).as(gf_calc_p_adj_start_date),
          col(settlement_flow_type).as(gf_float_rt_index_id),
          col(oper_flow_pymt_rcpt_mark_type).as(gf_cshflw_main_ctpty_role_type),
          col(legs_id).as(gf_trd_leg_id),
          col(phases_number).cast(IntegerType).as(gf_trd_phase_id),
          col(phase_flow_leg_sequential_id).cast(StringType).as(gf_trd_cshflw_bkwd_ofst_number),
          col(MUREX).as(gf_cshflw_typo_desc),
          when(col(fixed_var_flow_leg_mark_type) === F, lit(emptyString)).otherwise(col(fix_date)).as(gf_fixing_date),
          when(col(fixed_cost_amount) === zero, nullString).otherwise(col(fixed_cost_amount)).as(gf_ref_index_interest_rate_per),
          regexp_replace(when(col(fin_derv_oper_curr_flow_type).isin(IN,CP), col(fin_derv_int_notl_rgtry_date))
            .when(col(fin_derv_oper_curr_flow_type).isin(CL,PI), col(period_end_date))
            .otherwise(col(payment_date)), date_1900_01_01, emptyString).as(gf_adjusted_pmt_date),
          col(securities_nominal_amount).as(gf_ctrct_interest_rate_per),
          col(current_spread_amount).as(gf_int_rate_spread_margin_per),
          lit(nullString).as(gf_cshflw_amount),
          lit(nullString).as(g_cshflw_curncy_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  def tradeFlows(flowsMmo: DataFrame, flowsOtc: DataFrame, flowsDeposits: DataFrame): DataFrame = {
    try {
      flowsMmo.union(flowsOtc).union(flowsDeposits)
        .select(lit(MX).as(g_entific_id), lit(odate).cast(DateType).as(gf_cutoff_date), col(gf_origin_application_id), col(gf_front_sys_name),
          lit(nullString).cast(StringType).as(gf_trade_region_id), col(gf_front_sys_trd_id),
          when(col(gf_back_sys_trd_id) === emptyString, zero).when(col(gf_back_sys_trd_id) === " ", zero)
            .otherwise(col(gf_back_sys_trd_id)).as(gf_back_sys_trd_id), col(gf_local_contract_number_id),
          lit(nullString).cast(StringType).as(gf_srce_system_entity_id), lit(MX0074).as(g_entity_id), col(gf_trd_leg_id), col(gf_trd_phase_id).cast(StringType),
          lit(nullString).cast(DateType).as(gf_trd_date), lit(nullString).cast(StringType).as(gf_financial_asset_class_type),
          lit(nullString).cast(StringType).as(gf_fin_product_group_type), lit(nullString).cast(StringType).as(gf_fin_product_sub_group_type),
          lit(nullString).cast(StringType).as(gf_trd_flex_tmpl_id), lit(nullString).cast(DecimalType(eighteen,twelve)).as(gf_exchange_rate_amount),
          when(col(gf_trd_cshflw_bkwd_ofst_number).isNull, lit(one))
            .otherwise(col(gf_trd_cshflw_bkwd_ofst_number)).cast(DecimalType(teen,zero)).as(gf_trd_cshflw_bkwd_ofst_number),
          col(gf_cshflw_main_ctpty_role_type),
          lit(nullString).cast(StringType).as(gf_ctpty_source_system_id), lit(nullString).cast(StringType).as(gf_srce_system_counterparty_id),
          lit(nullString).cast(StringType).as(gf_counterparty_id), lit(nullString).cast(StringType).as(g_counterparty_id),
          lit(nullString).cast(StringType).as(gf_cstm_cshflw_ind_type), col(gf_cshflw_typo_desc),
          col(gf_cshflw_amount).cast(DecimalType(twentysix, six)).as(gf_cshflw_amount), col(g_cshflw_curncy_id).cast(StringType).as(g_cshflw_curncy_id),
          when(col(gf_calc_p_adj_end_date) === date_1900_01_01, emptyString).otherwise(col(gf_calc_p_adj_end_date)).cast(DateType).as(gf_calc_p_adj_end_date),
          when(col(gf_calc_p_adj_start_date).isNull, emptyString).otherwise(col(gf_calc_p_adj_start_date)).cast(DateType).as(gf_calc_p_adj_start_date),
          lit(nullString).cast(DecimalType(twentysix,six)).as(gf_cshflw_accrued_amount), lit(nullString).cast(StringType).as(gf_calc_basis_id),
          lit(nullString).cast(DecimalType(thirtyEight, zero)).as(gf_calc_p_days_number),
          lit(nullString).cast(DecimalType(twentysix,six)).as(gf_calc_p_days_fctr_number),
          col(g_interest_rate_type), col(gf_float_rt_index_id), lit(nullString).cast(StringType).as(gf_float_rt_index_tenor_id),
          col(gf_fixing_date).cast(DateType),
          when(col(gf_ref_index_interest_rate_per) === zero, nullString).when(col(gf_ref_index_interest_rate_per) === " ", nullString)
            .otherwise(col(gf_ref_index_interest_rate_per)).cast(DecimalType(twelve,nine)).as(gf_ref_index_interest_rate_per),
          lit(nullString).cast(DateType).as(gf_float_rt_fixing_p_end_date), lit(nullString).cast(DateType).as(gf_float_rt_fxg_p_start_date),
          when(col(gf_adjusted_pmt_date) === date_1900_01_01, lit(nullString)).otherwise(col(gf_adjusted_pmt_date)).cast(DateType).as(gf_adjusted_pmt_date),
          lit(nullString).cast(DecimalType(twentysix,six)).as(gf_cshflw_calc_notl_amount), lit(nullString).cast(StringType).as(g_cshflw_calc_notl_curncy_id),
          col(gf_ctrct_interest_rate_per).cast(DecimalType(twelve,nine)), lit(nullString).cast(DecimalType(twelve,nine)).as(gf_net_life_cap_per),
          lit(nullString).cast(DecimalType(twelve,nine)).as(gf_net_life_floor_per),lit(nullString).cast(DecimalType(twelve,four)).as(gf_float_rate_mult_number),
          col(gf_int_rate_spread_margin_per).cast(DecimalType(twelve,nine)), lit(nullString).cast(StringType).as(gf_cshflw_calc_formula_desc),
          lit(nullString).cast(StringType).as(gf_cmpnd_meth_typology_desc), lit(nullString).cast(DateType).as(gf_fx_rt_fixing_date),
          lit(nullString).cast(StringType).as(g_fx_rt_base_curncy_id), lit(nullString).cast(StringType).as(g_fx_rt_quote_curncy_id),
          lit(nullString).cast(StringType).as(gf_cshflw_subtype_desc), lit(nullString).cast(DecimalType(twenty,zero)).as(gf_cshflw_divd_calc_shr_number),
          lit(nullString).cast(DecimalType(twentysix,six)).as(gf_bps_sales_margin_amount),
          lit(nullString).cast(DecimalType(twentysix,six)).as(gf_cshflw_npv_amount),
          lit(nullString).cast(DecimalType(twelve,nine)).as(gf_cshflw_disc_fctr_per), lit(nullString).cast(DecimalType(five,two)).as(gf_transferred_per),
          lit(nullString).cast(DecimalType(twentysix,six)).as(gf_digital_rebate_amount), lit(nullString).cast(StringType).as(gf_related_trd_id),
          current_timestamp.as(gf_audit_date),
          lit(odate.replace("-", emptyString)).as(gf_odate_date_id),
          concat(lit(odate.replace("-", emptyString)), lit(Random.alphanumeric take 11 mkString)).as(gf_run_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
}
```

## NZTGTradeFlows

```scala
import Constants._
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.SparkSession
import scala.util.{Failure, Success, Try}
  val spark = SparkSession
        .builder()
        .appName("NztgTradeProd")
        .config("spark.some.config.option", "CDD")
        .getOrCreate()
  val odate = "2022-07-31"
  val hdate = "2022-07-29"
  val input = new InputReader(spark, hdate)
  val process = new Process(odate)
  // Process
  val catalogApp = input.readCatalogApp(t_ktny_catalog_values_taxonomy)
  val catalogFlowsType = input.catalogFlowsType()
  val tsiOperacion = input.readTsiOperacion(t_mdco_tsi001_operacion)
  val flagSwap = input.flagSwap(t_mrdc_liquidity_flows_over_swap)
  val flagOptions = input.flagOptions(t_mdco_liq_over_rate_options)
  val flowsSwpOpt = process.flowsSwpOpt(input.readLiquidityFlowsOverSwap(t_mrdc_liquidity_flows_over_swap, flagSwap),
     input.readLiqOverRateOptions(t_mdco_liq_over_rate_options, flagOptions), tsiOperacion)
  val flowsDeposOperacion = process.flowsDeposOperacion(input.readMoneyMarketDeposits(t_mdco_money_market_deposits), tsiOperacion)
  val flowsMmo = process.flowsMmo(input.readDerivadosMmooBis(t_mdco_derivados_mmoo_bis, catalogApp), flowsSwpOpt, catalogApp, catalogFlowsType)
  val flowsOtc = process.flowsOtc(input.readFinDervOtcBis(t_mdco_fin_derv_otc_bis, catalogApp), flowsSwpOpt, catalogApp, catalogFlowsType)
  val flowsDeposits = process.flowsDeposits(input.readInternalOperMurex(t_mdco_internal_oper_murex), flowsDeposOperacion, catalogFlowsType)
  // Trade_Flow
  val tradeFlows = process.tradeFlows(flowsMmo, flowsOtc, flowsDeposits)
```
