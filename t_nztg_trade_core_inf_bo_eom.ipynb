```scala
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.DataFrame
import org.apache.spark.sql.functions.{abs, broadcast, coalesce, col, concat, current_timestamp, lit, lpad, size, split, trim, upper, when}
import org.apache.spark.sql.types.{DateType, DecimalType, StringType, TimestampType}
import scala.util.Random

class Process(odate: String) {

 def fixedRent(rentaFijaOpeBis: DataFrame, tsiContraparte: DataFrame, tsiMarketOper: DataFrame,
                catalogApp: DataFrame, catalogPrsPcv: DataFrame, catalogOrf: DataFrame): DataFrame = {
    try {
      rentaFijaOpeBis.join(tsiContraparte, Seq(customer_corporate_id), left)
        .join(tsiMarketOper, Seq(operation_deal_front_star_id), left)
        .join(broadcast(catalogApp), rentaFijaOpeBis.col(source_application_id) === catalogApp.col(id), left)
        .join(broadcast(catalogPrsPcv), rentaFijaOpeBis.col(market_operation_type) === catalogPrsPcv.col(id), left)
        .join(broadcast(catalogOrf), rentaFijaOpeBis.col(securities_dealing_desk_name) === catalogOrf.col(id), left)
        .select(
          coalesce(catalogApp.col(value), rentaFijaOpeBis.col(source_application_id)).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), rentaFijaOpeBis.col(source_application_id)).as(gf_front_sys_name),
          when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero)).when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero))
            .otherwise(col(operation_deal_front_star_id)).as(gf_front_sys_trd_id),
          when(col(operation_deal_back_star_id) === emptyString, lit(ten_zero)).when(col(operation_deal_back_star_id) === zero_string, lit(ten_zero))
            .otherwise(col(operation_deal_back_star_id)).as(gf_back_sys_trd_id),
          col(market_operation_start_date).as(gf_trd_date), abs(col(initial_amount)).as(gf_initial_main_amount),
          when(col(market_operation_type).isin(productsFixedRent1:_*), lit(four_zero))
            .when(col(seller_branch_id).isNull, lit(four_zero))
              .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)).as(g_branch_id),
          lit(emptyString).alias(gf_trd_strategy_id),
          col(market_operation_settlement_date).as(gf_value_date),
          when(catalogPrsPcv.col(value).isNull, lit(X)).otherwise(catalogPrsPcv.col(value)).as(g_trade_position_type),
          col(isin_id).as(g_security_id), col(market_operation_type).as(gf_financial_product_id),
          lit(emptyString).alias(g_holding_product_id), lit("1611101010").alias(g_global_product_id),
          lit(tasa).alias(gf_undly_prd_typo_desc), col(origin_currency_type).as(g_currency_id),
          lit(emptyString).alias(gf_origin_trd_id),
          lit(nullString).cast(DateType).as(gf_trd_trmnt_original_date),
          tsiContraparte.col(counterpart_id).as(gf_srce_system_counterparty_id),
          tsiContraparte.col(counterparty_id).as(g_counterparty_id), lit(RDR).as(gf_ctpty_source_system_id),
          when(coalesce(tsiContraparte.col(customer_platform_id),
            rentaFijaOpeBis.col(customer_platform_id)) === lit(emptyString), lit(nullString))
            .otherwise(lpad(coalesce(tsiContraparte.col(customer_platform_id), rentaFijaOpeBis.col(customer_platform_id)), nine, zero_string))
                .as(g_customer_id),
          lit(emptyString).as(gf_ctpty_role_typo_desc),
          coalesce(catalogOrf.col(value), rentaFijaOpeBis.col(securities_dealing_desk_name)).as(gf_trading_desk_id),
          when(col(market_operation1_type) === E, lit(N)).when(col(market_operation1_type) === N, lit(Y)).otherwise(lit(N)).as(g_internal_operation_ind_type),
          lit(Y).as(gf_trd_cpt_cnsptn_pr_ind_type), lit(emptyString).as(gf_break_clause_date_desc),
          lit(bilt).as(g_mic_id), lit(insert).as(gf_trd_ori_event_typo_desc), col(market_operation_start_date).as(gf_trd_change_effective_date),
          lit(S).as(g_spot_forward_type), col(next_coupon_cutoff_date).as(gf_interest_rate_change_date), lit(emptyString).as(g_renewal_type),
          lit(N).as(gf_deriv_setl_mkt_ind_type), lit(N).as(gf_clear_cust_op_ind_type), lit(N).as(gf_derv_net_dpst_ind_type),
          coalesce(catalogOrf.col(value), rentaFijaOpeBis.col(securities_dealing_desk_name)).as(gf_management_portfolio_id),
          col(market_operation_expiry_date).as(gf_trd_adj_trmnt_date), lit(MX0074).as(g_entity_id),
          concat(coalesce(catalogApp.col(value), when(rentaFijaOpeBis.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(rentaFijaOpeBis.col(source_application_id))),
            when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero)).otherwise(col(operation_deal_front_star_id)))
            .as(g_contract_id),
          when(col(market_operation_type).isin(productsFixedRent1:_*), concat(col(var_coverage_secty_mark_type), col(isin_id)))
            .when(col(market_operation_type).isin(productsFixedRent2:_*),
              when(col(operation_deal_back_star_id) === emptyString, lit(ten_zero))
                .when(col(operation_deal_back_star_id) === zero_string, lit(ten_zero))
                .otherwise(col(operation_deal_back_star_id)))
            .otherwise(when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero))
              .otherwise(col(operation_deal_front_star_id))).as(gf_local_contract_number_id),
          col(current_commitment_amount).cast(DecimalType(twenty_six, six)).as(gf_fin_trade_price_amount),
          when(col(market_operation_type).isin(productsFixedRent3:_*) && col(securities_dealing_desk_name) === lit(COA), lit(oneFourOne))
           .when(col(market_operation_type).isin(productsFixedRent4:_*) && col(securities_dealing_desk_name) === lit(DGM), lit(zeroFiveZero))
            .when(col(market_operation_type).isin(productsFixedRent5:_*) && col(securities_dealing_desk_name) === lit(DGM), lit(zeroNineSix))
             .when(col(market_operation_type) === lit(PRB) && col(securities_dealing_desk_name) === lit(DGM), lit(zeroOneZero))
              .when(col(market_operation_type) === lit(VR) && col(securities_dealing_desk_name) === lit(COA), lit(oneEightOne))
               .when(col(market_operation_type) === lit(GET) && col(securities_dealing_desk_name) === lit(DGM), lit(zeroFiveZero))
                .otherwise(lit(oneOneZero))
                 .as(g_gl_classification_type),
          lit(emptyString).as(g_trade_mirror_ind_type),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def fx(fxDf: DataFrame, economicsDataSwapsDivFwd: DataFrame, tsiContraparte: DataFrame,
         tsiContraparteRefactored: DataFrame, tsiFwdDiv: DataFrame,
         originOperations: DataFrame, tsiOperacion: DataFrame, catalogApp: DataFrame, catalogPrsPcv: DataFrame,
         catalogSuby: DataFrame, catalogHoldingProduct: DataFrame, originOperationTrdId: DataFrame): DataFrame = {
    try {
      fxDf.join(economicsDataSwapsDivFwd, Seq(operation_deal_front_star_id, portfolio_type, counterparty_id), "inner")
        .join(tsiContraparte, Seq(customer_platform_id, customer_corporate_id), left)
        .join(tsiFwdDiv, Seq(operation_deal_front_star_id, portfolio_type, counterparty_id), left)
        .join(originOperations, fxDf.col(operation_deal_front_star_id) === originOperations.col(update_gen_market_oper_number), left)
        .join(tsiOperacion, Seq(operation_deal_front_star_id, portfolio_type, counterparty_id), left)
        .join(tsiContraparteRefactored, Seq(counterparty_id), left)
        .join(broadcast(catalogApp), fxDf.col(source_application_id) === catalogApp.col(id), left)
        .join(broadcast(catalogPrsPcv), fxDf.col(operation_nature_type) === catalogPrsPcv.col(id), left)
        .join(broadcast(catalogSuby), fxDf.col(subproduct_type) === catalogSuby.col(id), left)
        .join(broadcast(catalogHoldingProduct), col(subproduct_type) === col(market_type), left)
        .join(originOperationTrdId, Seq(operation_deal_front_star_id), left)
        .select(coalesce(catalogApp.col(value), fxDf.col(source_application_id)).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), fxDf.col(source_application_id)).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id),
          col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          col(contract_register_date).as(gf_trd_date), abs(col(market_oper_recv_notional_amount)).as(gf_initial_main_amount),
          when(col(market_operation_type).isin(productsFixedRent1:_*), lit(four_zero))
            .when(col(seller_branch_id).isNull, lit(four_zero))
            .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)).as(g_branch_id),
          when(col(product_grouped_id).isNull, col(portfolio_operation_strategy_id)).otherwise(col(product_grouped_id)).as(gf_trd_strategy_id),
          col(value_date).as(gf_value_date),
          when(catalogPrsPcv.col(value).isNull,
            lit(X)).otherwise(catalogPrsPcv.col(value)).as(g_trade_position_type),
          lit(emptyString).as(g_security_id), upper(col(subproduct_type)).as(gf_financial_product_id),
          lit(emptyString).as(g_holding_product_id), col(cd_holding).as(g_global_product_id),
          coalesce(catalogSuby.col(value), fxDf.col(subproduct_type)).as(gf_undly_prd_typo_desc), col(operation_currency_id).as(g_currency_id),
          coalesce(col(first_deal), lit(emptyString)).as(gf_origin_trd_id),
          coalesce(col(market_operation_expiry_date), lit(nullString).cast(DateType)).as(gf_trd_trmnt_original_date),
          tsiContraparte.col(counterpart_id).as(gf_srce_system_counterparty_id),
          tsiContraparte.col(counterparty_id).as(g_counterparty_id), lit(RDR).as(gf_ctpty_source_system_id),
          when(coalesce(col(customer_platform_id), col(customer_platform_id_refactored)) === lit(emptyString), lit(nullString))
            .otherwise(lpad(coalesce(col(customer_platform_id), col(customer_platform_id_refactored)), nine, zero_string))
            .as(g_customer_id),
          lit(emptyString).as(gf_ctpty_role_typo_desc),
          lit(split(col(object_operation_name), "-")(size(split(col(object_operation_name), "-")) - 1)).as(gf_trading_desk_id),
          when(col(trading_mark_type) === EXTERNA, lit(N)).when(col(trading_mark_type) === INTERNA, lit(Y))
            .otherwise(lit(emptyString)).as(g_internal_operation_ind_type),
          lit(Y).as(gf_trd_cpt_cnsptn_pr_ind_type), economicsDataSwapsDivFwd.col(break_clause1_date).as(gf_break_clause_date_desc), lit(XOFF).as(g_mic_id),
          coalesce(when(col(source_application_id) === MX, col(market_operation_expiry_type)), lit(INSERT)).as(gf_trd_ori_event_typo_desc),
          coalesce(when(col(source_application_id) === MX, col(market_operation_date)), col(contract_register_date)).as(gf_trd_change_effective_date),
          when(col(subproduct_type).isin(ffxpr1: _*), F).when(col(subproduct_type) === ffxpr2, S)
            .when(col(subproduct_type) === ffxpr3 && economicsDataSwapsDivFwd.col(operation_days_number) <= 4, S)
            .otherwise(coalesce(when(economicsDataSwapsDivFwd.col(operation_days_number) <= 2, S), lit(F))).as(g_spot_forward_type),
          lit(emptyString).as(gf_interest_rate_change_date), lit(emptyString).as(g_renewal_type),
          lit(N).as(gf_deriv_setl_mkt_ind_type), lit(N).as(gf_clear_cust_op_ind_type), lit(N).as(gf_derv_net_dpst_ind_type),
          fxDf.col(portfolio_type).as(gf_management_portfolio_id),
          col(expiration_date).as(gf_trd_adj_trmnt_date), lit(MX0074).as(g_entity_id),
          concat(coalesce(catalogApp.col(value), when(fxDf.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(fxDf.col(source_application_id))),
            trim(col(operation_deal_front_star_id))).as(g_contract_id),
          col(operation_deal_front_star_id).as(gf_local_contract_number_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_fin_trade_price_amount),
          when(fxDf.col(object_operation_name) === lit(NEGOTATION), lit(zeroFiveZero))
           .when(fxDf.col(object_operation_name) === lit(SPECULATIONTRADING), lit(zeroFiveZero))
            .when(fxDf.col(object_operation_name) === lit(COVERAGE), lit(oneFourOne))
             .otherwise(fxDf.col(object_operation_name))
              .as(g_gl_classification_type),
            when(col(offc_grp_intrnl_oper_img_type) === lit(Y), lit(N))
              .when(col(offc_grp_intrnl_oper_img_type).isNull, lit(N))
               .when(col(offc_grp_intrnl_oper_img_type) === lit(N), lit(Y))
                .otherwise(col(offc_grp_intrnl_oper_img_type))
                 .as(g_trade_mirror_ind_type),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def previousOrganizedMarkets(derivadosMmooBis: DataFrame, tsiMarketOper: DataFrame, tsiContraparte: DataFrame, tsiOperacion: DataFrame,
                               catalogApp: DataFrame, catalogSuby: DataFrame, catalogHoldingProduct: DataFrame, originOperationTrdId: DataFrame): DataFrame = {
    try {
      derivadosMmooBis.join(tsiMarketOper, Seq(operation_deal_front_star_id), left)
        .join(tsiContraparte, Seq(customer_platform_id, customer_corporate_id), left)
        .join(tsiOperacion, Seq(operation_deal_front_star_id, counterparty_id), left)
        .join(broadcast(catalogApp), derivadosMmooBis.col(source_application_id) === catalogApp.col(id), left)
        .join(broadcast(catalogSuby), derivadosMmooBis.col(underline_asset_value_type_id) === catalogSuby.col(id), left)
        .join(broadcast(catalogHoldingProduct), derivadosMmooBis.col(product_type) === catalogHoldingProduct.col(market_type), left)
        .join(originOperationTrdId, Seq(operation_deal_front_star_id), left)
        .select(coalesce(catalogApp.col(value), col(source_application_id)).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), col(source_application_id)).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id), col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          col(contract_register_date).as(gf_trd_date),
          abs(when(col(trading_mark_type) === V, col(market_oper_pay_notional_amount))
            .otherwise(col(market_oper_recv_notional_amount))).as(gf_initial_main_amount),
          when(col(seller_branch_id).isNull, lit(four_zero))
            .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string))
            .as(g_branch_id),
          col(product_grouped_id).as(gf_trd_strategy_id), col(value_date).as(gf_value_date),
          when(col(product_type).isin(IRS, IRC),
            col(market_operation_type)).otherwise(col(trading_mark_type)).as(g_trade_position_type),
          lit(emptyString).as(g_security_id), upper(col(product_type)).as(gf_financial_product_id), lit(emptyString).as(g_holding_product_id),
          col(cd_holding).as(g_global_product_id), coalesce(catalogSuby.col(value),
            derivadosMmooBis.col(underline_asset_value_type_id)).as(gf_undly_prd_typo_desc),
          when(col(trading_mark_type) === C, col(market_oper_recv_notl_curncy_id)).otherwise(col(counterparty_pay_notl_curncy_id)).as(g_currency_id),
          coalesce(col(first_deal), lit(emptyString)).as(gf_origin_trd_id),
          coalesce(col(market_operation_expiry_date), lit(nullString).cast(DateType)).as(gf_trd_trmnt_original_date),
          tsiContraparte.col(counterpart_id).as(gf_srce_system_counterparty_id),
          tsiContraparte.col(counterparty_id).as(g_counterparty_id),
          lit(RDR).as(gf_ctpty_source_system_id),
          when(col(customer_platform_id) === lit(emptyString), lit(nullString))
            .otherwise(lpad(col(customer_platform_id), nine, zero_string)).as(g_customer_id),
          lit(emptyString).as(gf_ctpty_role_typo_desc),
          lit(emptyString).as(gf_trading_desk_id), lit(N).as(g_internal_operation_ind_type), lit(Y).as(gf_trd_cpt_cnsptn_pr_ind_type),
          lit(emptyString).as(gf_break_clause_date_desc),
          when(col(source_application_id).isin(MUREX, CORAL) && col(financial_asset_group_type) === MEXDER, XEMD).otherwise(lit(BILT)).as(g_mic_id),
          when(coalesce(catalogApp.col(value), derivadosMmooBis.col(source_application_id)) === MUREX && col(market_operation_expiry_type).isNull, RPL_M)
            .when(coalesce(catalogApp.col(value), derivadosMmooBis.col(source_application_id)) === ST && col(market_operation_expiry_type) === RPL_M, INSERT)
            .otherwise(lit(INSERT)).as(gf_trd_ori_event_typo_desc),
          when(tsiMarketOper.col(market_operation_date).isNull, derivadosMmooBis.col(contract_register_date))
            .otherwise(tsiMarketOper.col(market_operation_date)).as(gf_trd_change_effective_date), lit(F).as(g_spot_forward_type),
          when(col(trading_mark_type) === V, col(pay_coupon_end_date)).otherwise(col(receive_coupon_payment_end_date)).as(gf_interest_rate_change_date),
          lit(emptyString).as(g_renewal_type), lit(N).as(gf_deriv_setl_mkt_ind_type), lit(N).as(gf_clear_cust_op_ind_type),
          lit(N).as(gf_derv_net_dpst_ind_type), lit(emptyString).as(gf_management_portfolio_id), col(expiration_date).as(gf_trd_adj_trmnt_date),
          lit(MX0074).as(g_entity_id), concat(trim(coalesce(catalogApp.col(value),
            when(derivadosMmooBis.col(source_application_id) === emptyString, BGM_BIS)
              .otherwise(derivadosMmooBis.col(source_application_id)))),
            when(trim(col(operation_deal_front_star_id)) === zero_string, trim(col(operation_deal_back_star_id))).
              otherwise(col(operation_deal_front_star_id))).as(g_contract_id),
          when(trim(col(operation_deal_front_star_id)) === zero_string, trim(col(operation_deal_back_star_id))).
            otherwise(col(operation_deal_front_star_id)).as(gf_local_contract_number_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_fin_trade_price_amount),
          lit(oneEightOne).as(g_gl_classification_type),
          when(col(offc_grp_intrnl_oper_img_type) === lit(Y), lit(N))
            .when(col(offc_grp_intrnl_oper_img_type).isNull, lit(N))
             .when(col(offc_grp_intrnl_oper_img_type) === lit(N), lit(Y))
              .otherwise(col(offc_grp_intrnl_oper_img_type))
               .as(g_trade_mirror_ind_type),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def organizedMarkets(previousOrganizedMarkets: DataFrame, catalogPrsPcv: DataFrame): DataFrame = {
    try {
      previousOrganizedMarkets.join(broadcast(catalogPrsPcv), col(g_trade_position_type) === col(id), left)
        .select(
          when(col(gf_origin_application_id) === emptyString, BGM_BIS)
            .otherwise(col(gf_origin_application_id)).as(gf_origin_application_id),
          when(col(gf_front_sys_name) === emptyString, BGM_BIS)
            .otherwise(col(gf_front_sys_name)).as(gf_front_sys_name),
          col(gf_front_sys_trd_id),
          col(gf_back_sys_trd_id),
          col(gf_trd_date),
          col(gf_initial_main_amount),
          col(g_branch_id),
          col(gf_trd_strategy_id),
          col(gf_value_date),
          coalesce(col(value), col(g_trade_position_type)).as(g_trade_position_type),
          col(g_security_id),
          col(gf_financial_product_id),
          col(g_holding_product_id),
          col(g_global_product_id),
          col(gf_undly_prd_typo_desc),
          col(g_currency_id),
          col(gf_origin_trd_id),
          col(gf_trd_trmnt_original_date),
          col(gf_srce_system_counterparty_id),
          col(g_counterparty_id),
          col(gf_ctpty_source_system_id),
          col(g_customer_id),
          col(gf_ctpty_role_typo_desc),
          col(gf_trading_desk_id),
          col(g_internal_operation_ind_type),
          col(gf_trd_cpt_cnsptn_pr_ind_type),
          col(gf_break_clause_date_desc),
          col(g_mic_id),
          col(gf_trd_ori_event_typo_desc),
          col(gf_trd_change_effective_date),
          col(g_spot_forward_type),
          col(gf_interest_rate_change_date),
          col(g_renewal_type),
          col(gf_deriv_setl_mkt_ind_type),
          col(gf_clear_cust_op_ind_type),
          col(gf_derv_net_dpst_ind_type),
          col(gf_management_portfolio_id),
          col(gf_trd_adj_trmnt_date),
          col(g_entity_id),
          col(g_contract_id),
          col(gf_local_contract_number_id),
          col(gf_fin_trade_price_amount),
          col(g_gl_classification_type),
          col(g_trade_mirror_ind_type),
          col(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def previousOtc(finDervOtcBis: DataFrame, economicsDataSwapsDivFwd: DataFrame, cbOperations: DataFrame, tsiContraparte: DataFrame,
                  originOperations: DataFrame, catalogApp: DataFrame, catalogSuby: DataFrame, originOperationTrdId: DataFrame): DataFrame = {
    try {
      finDervOtcBis.join(economicsDataSwapsDivFwd, Seq(operation_deal_front_star_id, portfolio_type, counterparty_id), left)
        .join(cbOperations, Seq(operation_deal_front_star_id, portfolio_type, counterparty_id), left)
        .join(tsiContraparte, Seq(customer_platform_id, customer_corporate_id), left)
        .filter((col(subproduct_type).as(gf_financial_product_id)
          .isin(IRC, IRS) && col(market_oper_pay_notional_amount) =!= 0) or !col(subproduct_type)
          .as(gf_financial_product_id).isin(IRC, IRS))
        .join(originOperations.select(col(murex_operation_id), col(market_operation_expiry_type), col(market_operation_date),
          col(update_gen_market_oper_number).as(operation_deal_front_star_id)), Seq(operation_deal_front_star_id), left)
        .join(broadcast(catalogApp), finDervOtcBis.col(source_application_id) === catalogApp.col(id), left)
        .join(broadcast(catalogSuby), finDervOtcBis.col(subproduct_type) === catalogSuby.col(id), left)
        .join(originOperationTrdId, Seq(operation_deal_front_star_id), left)
        .select(
          coalesce(catalogApp.col(value), finDervOtcBis.col(source_application_id)).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), finDervOtcBis.col(source_application_id)).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id), col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          col(contract_register_date).as(gf_trd_date),
          abs(when(col(subproduct_type).isin(IRC, IRS) && col(gl_account_type) === D && col(gl_account_id).startsWith("6"), col(current_balance_amount))
            .when(finDervOtcBis.col(market_oper_pay_notional_amount) =!= number_zero, finDervOtcBis.col(market_oper_pay_notional_amount))
            .otherwise(col(market_oper_recv_notional_amount))).as(gf_initial_main_amount),
          when(col(seller_branch_id).isNull, lit(four_zero))
            .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string))
            .as(g_branch_id),
          col(product_grouped_id).as(gf_trd_strategy_id), col(value_date).as(gf_value_date),
          when(cbOperations.col(market_operation_type).isNotNull, cbOperations.col(market_operation_type))
            .when(col(trading_mark_type) === lit(emptyString), lit(X))
            .otherwise(col(trading_mark_type)).as(g_trade_position_type),
          lit(emptyString).as(g_security_id), upper(col(subproduct_type)).as(gf_financial_product_id),
          lit(emptyString).as(g_holding_product_id), col(cd_holding).as(g_global_product_id),
          coalesce(catalogSuby.col(value), finDervOtcBis.col(subproduct_type)).as(gf_undly_prd_typo_desc),
          when(col(currency_id).isNull || trim(col(currency_id)) === emptyString, col(currency1_id)).otherwise(col(currency_id)).as(g_currency_id),
          coalesce(col(first_deal), lit(emptyString)).as(gf_origin_trd_id),
          coalesce(col(market_operation_expiry_date), lit(nullString).cast(DateType)).as(gf_trd_trmnt_original_date),
          tsiContraparte.col(counterparty_id).as(g_counterparty_id),
          tsiContraparte.col(counterpart_id).as(gf_srce_system_counterparty_id),
          lit(RDR).as(gf_ctpty_source_system_id),
          when(col(customer_platform_id) === lit(emptyString), lit(nullString))
            .otherwise(lpad(col(customer_platform_id), nine, zero_string)).as(g_customer_id),
          lit(emptyString).as(gf_ctpty_role_typo_desc), col(object_operation_name).as(gf_trading_desk_id),
          when(col(contract_operation_type) === EXTERNA, N).when(col(contract_operation_type) === INTERNA, Y)
            .otherwise(emptyString).as(g_internal_operation_ind_type), lit(Y).as(gf_trd_cpt_cnsptn_pr_ind_type),
          col(break_clause1_date).as(gf_break_clause_date_desc), lit(BILT).alias(g_mic_id),
          coalesce(when(col(source_application_id) === MX, col(market_operation_expiry_type)), lit(INSERT)).as(gf_trd_ori_event_typo_desc),
          coalesce(when(col(source_application_id) === MX, col(market_operation_date)),
            col(contract_register_date)).alias(gf_trd_change_effective_date), lit(F).as(g_spot_forward_type),
          when(col(trading_mark_type) === V, col(pay_coupon_end_date))
            .otherwise(col(receive_coupon_payment_end_date)).as(gf_interest_rate_change_date),
          lit(emptyString).as(g_renewal_type), lit(N).as(gf_deriv_setl_mkt_ind_type), lit(N).as(gf_clear_cust_op_ind_type),
          lit(N).as(gf_derv_net_dpst_ind_type), col(portfolio_type).as(gf_management_portfolio_id),
          col(expiration_date).as(gf_trd_adj_trmnt_date), lit(MX0074).alias(g_entity_id),
          concat(trim(coalesce(catalogApp.col(value), when(finDervOtcBis.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(finDervOtcBis.col(source_application_id)))), trim(col(operation_deal_front_star_id))).as(g_contract_id),
          trim(col(operation_deal_front_star_id)).as(gf_local_contract_number_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_fin_trade_price_amount),
          when (col(object_operation_name) === lit(NEGOTATION), lit(zeroFiveZero))
            .when(col(object_operation_name) === lit(COVERAGE), lit(oneFourOne))
             .otherwise(col(object_operation_name))
              .as(g_gl_classification_type),
          lit(emptyString).as(g_trade_mirror_ind_type),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def otc(previousOtc: DataFrame, catalogPrsPcv: DataFrame): DataFrame = {
    try {
      previousOtc.join(broadcast(catalogPrsPcv), col(g_trade_position_type) === col(id), left)
        .select(
          when(col(gf_origin_application_id) === emptyString, BGM_BIS)
            .otherwise(col(gf_origin_application_id)).as(gf_origin_application_id),
          when(col(gf_front_sys_name) === emptyString, BGM_BIS)
            .otherwise(col(gf_front_sys_name)).as(gf_front_sys_name),
          col(gf_front_sys_trd_id),
          col(gf_back_sys_trd_id),
          col(gf_trd_date),
          col(gf_initial_main_amount),
          col(g_branch_id),
          col(gf_trd_strategy_id),
          col(gf_value_date),
          coalesce(col(value), col(g_trade_position_type)).as(g_trade_position_type),
          col(g_security_id),
          col(gf_financial_product_id),
          col(g_holding_product_id),
          col(g_global_product_id),
          col(gf_undly_prd_typo_desc),
          col(g_currency_id),
          col(gf_origin_trd_id),
          col(gf_trd_trmnt_original_date),
          col(gf_srce_system_counterparty_id),
          col(g_counterparty_id),
          col(gf_ctpty_source_system_id),
          col(g_customer_id),
          col(gf_ctpty_role_typo_desc),
          col(gf_trading_desk_id),
          col(g_internal_operation_ind_type),
          col(gf_trd_cpt_cnsptn_pr_ind_type),
          col(gf_break_clause_date_desc),
          col(g_mic_id),
          col(gf_trd_ori_event_typo_desc),
          col(gf_trd_change_effective_date),
          col(g_spot_forward_type),
          col(gf_interest_rate_change_date),
          col(g_renewal_type),
          col(gf_deriv_setl_mkt_ind_type),
          col(gf_clear_cust_op_ind_type),
          col(gf_derv_net_dpst_ind_type),
          col(gf_management_portfolio_id),
          col(gf_trd_adj_trmnt_date),
          col(g_entity_id),
          col(g_contract_id),
          col(gf_local_contract_number_id),
          col(gf_fin_trade_price_amount),
          col(g_gl_classification_type),
          col(g_trade_mirror_ind_type),
          col(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def deposits(internalOperMurex: DataFrame, tsiContraparte: DataFrame,
               tsiMarketOper: DataFrame, mdcoMoneyMktBis: DataFrame): DataFrame = {
    try {
      internalOperMurex
        .join(tsiContraparte, col(app_counterparty_id) === col(counterpart_id), left)
        .join(tsiMarketOper, Seq(operation_deal_front_star_id), left)
        .join(mdcoMoneyMktBis, Seq(operation_deal_front_star_id), left)
        .select(
          lit(MUREX).as(gf_origin_application_id),
          lit(MUREX).as(gf_front_sys_name),
          col(operation_deal_front_star_id).cast(StringType).as(gf_front_sys_trd_id),
          lit(emptyString).as(gf_back_sys_trd_id),
          col(contract_register_date).as(gf_trd_date),
          abs(col(fin_instm_nominal_amount)).as(gf_initial_main_amount),
          when(col(seller_branch_id).isNull, lit(four_zero))
            .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string))
            .as(g_branch_id),
          col(product_grouped_id).as(gf_trd_strategy_id),
          col(system_operation_register_date).as(gf_value_date),
          col(transaction_type).as(g_trade_position_type),
          lit(emptyString).as(g_security_id),
          lit(DEPOSITS).as(gf_financial_product_id),
          lit(emptyString).as(g_holding_product_id),
          lit("1610111010").as(g_global_product_id),
          lit(emptyString).as(gf_undly_prd_typo_desc),
          col(nominal_currency_id).as(g_currency_id),
          coalesce(when(col(market_operation_expiry_type).isNotNull, col(operation_deal_front_star_id)),
            lit(emptyString)).as(gf_origin_trd_id),
          lit(nullString).cast(DateType).as(gf_trd_trmnt_original_date),
          tsiContraparte.col(counterpart_id).as(gf_srce_system_counterparty_id),
          tsiContraparte.col(counterparty_id).as(g_counterparty_id),
          lit(RDR).as(gf_ctpty_source_system_id),
          when(col(customer_platform_id) === lit(emptyString), lit(nullString)).otherwise(lpad(col(customer_platform_id), nine, zero_string))
            .as(g_customer_id),
          lit(emptyString).as(gf_ctpty_role_typo_desc),
          lit(emptyString).as(gf_trading_desk_id),
          col(int_ext_operative_mark_type).as(g_internal_operation_ind_type),
          lit(Y).as(gf_trd_cpt_cnsptn_pr_ind_type),
          lit(emptyString).as(gf_break_clause_date_desc),
          lit(XOFF).as(g_mic_id),
          coalesce(col(market_operation_expiry_type), lit(INSERT)).as(gf_trd_ori_event_typo_desc),
          coalesce(col(market_operation_date), col(contract_register_date)).as(gf_trd_change_effective_date),
          lit(F).as(g_spot_forward_type),
          col(next_interest_pymt_calc_date).as(gf_interest_rate_change_date),
          lit(emptyString).as(g_renewal_type),
          lit(N).as(gf_deriv_setl_mkt_ind_type),
          lit(N).as(gf_clear_cust_op_ind_type),
          lit(N).as(gf_derv_net_dpst_ind_type),
          col(portfolio_investment_name).as(gf_management_portfolio_id),
          col(expiration_date).as(gf_trd_adj_trmnt_date),
          lit(MX0074).as(g_entity_id),
          concat(col(operation_deal_front_star_id)).as(g_contract_id),
          concat(col(operation_deal_front_star_id)).as(gf_local_contract_number_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_fin_trade_price_amount),
          lit(oneEightOne).as(g_gl_classification_type),
          lit(emptyString).as(g_trade_mirror_ind_type),
          lit(N).as(g_triparty_trade_ind_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  
  def tradeCore(fixedRent: DataFrame, fx: DataFrame, organizedMarket: DataFrame, otc: DataFrame, deposits: DataFrame, catalogCurrency: DataFrame): DataFrame = {
    try {
      fixedRent.union(fx).union(organizedMarket).union(otc).union(deposits)
        .join(broadcast(catalogCurrency), col(g_currency_id) === catalogCurrency.col(id), left)
        .select(lit(country).as(g_entific_id), col(gf_origin_application_id), col(gf_front_sys_name), lit(emptyString).as(gf_trade_region_id),
          col(gf_front_sys_trd_id), col(gf_back_sys_trd_id), col(g_contract_id),
          concat(col(g_branch_id), col(gf_local_contract_number_id)).as(gf_local_contract_number_id),
          col(gf_trd_date),
          lit(nullString).cast(StringType).as(gf_trd_info_period_typo_desc), lit(odate).cast(DateType).as(gf_cutoff_date),
          col(gf_initial_main_amount).cast(DecimalType(twenty_six, six)), lit(nullString).cast(StringType).as(g_trd_notl_curncy_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_curr_notl_amount), lit(nullString).cast(StringType).as(g_trd_curr_notl_curncy_id),
          lit(nullString).cast(DecimalType(thirty_six, sixteen)).as(gf_total_nominal_eur_amount),
          lit(nullString).cast(StringType).as(g_countervalued_euro_curncy_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_exchange_rate_euro_amount), lit(nullString).cast(StringType).as(g_countervalued_currency_id),
          lit(nullString).cast(DecimalType(twelve, four)).as(gf_trd_notl_mult_number),
          lit(nullString).cast(DecimalType(twenty_six, number_zero)).as(gf_trd_prdt_qty_number),
          lit(nullString).cast(DecimalType(twelve, four)).as(gf_contract_multiplier_number), lit(nullString).cast(StringType).as(gf_srce_system_entity_id),
          lit(nullString).cast(StringType).as(gf_entity_name), col(g_entity_id), lit(nullString).cast(StringType).as(gf_trd_closing_entity_id),
          col(g_branch_id), col(gf_trd_strategy_id),
          lit(nullString).cast(StringType).as(gf_telemaco_trade_id),
          lit(nullString).cast(StringType).as(gf_trd_previous_strategy_id),
          lit(nullString).cast(DateType).as(gf_front_sys_trd_info_date), lit(nullString).cast(TimestampType).as(gf_front_sys_trd_info_tme_date),
          lit(nullString).cast(DateType).as(gf_trd_unadj_eff_date), lit(nullString).cast(StringType).as(gf_trd_eff_date_bdc_typo_desc),
          col(gf_value_date), lit(nullString).cast(DateType).as(gf_trd_unadj_trmnt_date), lit(nullString).cast(StringType).as(gf_trd_trmnt_dt_bdc_typo_desc),
          col(gf_trd_adj_trmnt_date), lit(nullString).cast(DateType).as(gf_trd_effective_end_date),
          col(gf_trd_trmnt_original_date).cast(DateType),
          lit(nullString).cast(StringType).as(g_bought_sold_position_type), col(g_trade_position_type),
          lit(nullString).cast(StringType).as(gf_financial_asset_class_type),
          lit(nullString).cast(StringType).as(gf_fin_product_group_type), lit(nullString).cast(StringType).as(gf_fin_product_sub_group_type),
          lit(nullString).cast(StringType).as(gf_trd_flex_tmpl_id), lit(nullString).cast(StringType).as(gf_trd_flex_tmpl_desc),
          lit(nullString).cast(StringType).as(gf_fin_product_typo_desc), lit(nullString).cast(StringType).as(gf_financial_product_desc),
          col(g_security_id), lit(nullString).cast(StringType).as(gf_secur_label_desc), col(gf_financial_product_id), lit(xHolding).as(g_holding_product_id),
          col(g_global_product_id), lit(nullString).cast(StringType).as(gf_trd_product_assn_id), col(gf_undly_prd_typo_desc),
          coalesce(catalogCurrency.col(value), col(g_currency_id)).as(g_currency_id),
          lit(nullString).cast(DecimalType(seventeen, twelve)).as(gf_fx_spot_exch_rate_amount),
          lit(nullString).cast(StringType).as(gf_trd_origin_application_id),
          lit(nullString).cast(StringType).as(gf_source_system_desc), lit(nullString).cast(StringType).as(gf_external_trd_id),
          lit(nullString).cast(StringType).as(g_murex_global_trd_id), lit(nullString).cast(StringType).as(gf_trd_fee_id),
          lit(nullString).cast(StringType).as(gf_mexder_usi_trd_id), col(gf_origin_trd_id), lit(nullString).cast(StringType).as(gf_creator_trd_id),
          lit(nullString).cast(StringType).as(gf_trading_venue_id), lit(nullString).cast(StringType).as(gf_trading_venue_trd_id),
          lit(nullString).cast(StringType).as(gf_trd_venue_far_leg_trd_id), lit(nullString).cast(StringType).as(gf_trd_last_trd_plat_id),
          lit(nullString).cast(StringType).as(gf_last_trd_plat_trd_id), lit(nullString).cast(StringType).as(gf_intn_to_clear_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_clearing_orgn_trd_id), lit(nullString).cast(StringType).as(gf_far_leg_trd_id),
          lit(nullString).cast(StringType).as(gf_structured_note_lot_id), lit(nullString).cast(StringType).as(g_markit_icelink_trd_id),
          lit(nullString).cast(StringType).as(gf_previous_trd_id),
          lit(nullString).cast(StringType).as(gf_rollover_new_trade_id),
          lit(nullString).cast(StringType).as(gf_rollover_parent_trade_id),
          lit(nullString).cast(StringType).as(gf_mirr_trd_id),
          col(g_trade_mirror_ind_type),
          lit(nullString).cast(StringType).as(gf_mirr_far_leg_intrnl_trd_id),
          lit(nullString).cast(StringType).as(gf_trd_mirr_book_id),
          lit(nullString).cast(StringType).as(gf_trn_mirr_trd_ent_id),
          lit(nullString).cast(StringType).as(gf_front_sys_trd_int_assn_id),
          lit(nullString).cast(StringType).as(gf_trd_assoc_retail_trd_id), lit(nullString).cast(StringType).as(gf_trd_package_id),
          lit(nullString).cast(StringType).as(gf_package_trd_creator_id), lit(nullString).cast(StringType).as(gf_origin_trd_package_id),
          lit(nullString).cast(StringType).as(gf_trd_origin_option_id), lit(nullString).cast(DecimalType(teen, number_zero)).as(gf_trd_ver_number),
          lit(nullString).cast(DecimalType(teen, number_zero)).as(gf_trd_contract_ver_number), lit(nullString).cast(StringType).as(gf_creator_trd_contract_id),
          lit(nullString).cast(StringType).as(gf_current_trd_contract_id), lit(nullString).cast(StringType).as(gf_root_trd_contract_id),
          lit(nullString).cast(StringType).as(gf_gl_hedge_assn_id), lit(nullString).cast(StringType).as(gf_operative_hedge_assn_id),
          lit(nullString).cast(StringType).as(g_unique_swap_id), lit(nullString).cast(StringType).as(g_unique_trd_id),
          lit(nullString).cast(StringType).as(g_prior_unique_trd_id), lit(nullString).cast(StringType).as(gf_trd_secur_exch_ind_type),
          lit(nullString).cast(StringType).as(gf_block_trd_id), lit(nullString).cast(StringType).as(gf_trd_alloc_status_typo_desc),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_trd_allocated_notional_per), lit(nullString).cast(StringType).as(gf_reprocess_sys_ind_type),
          col(gf_ctpty_source_system_id), col(gf_srce_system_counterparty_id), lit(nullString).cast(StringType).as(gf_counterparty_id), col(g_counterparty_id),
          col(g_customer_id), col(gf_ctpty_role_typo_desc), lit(nullString).cast(StringType).as(gf_counterparty_desc),
          lit(nullString).cast(StringType).as(gf_trd_ctpty_pfolio_id), lit(nullString).cast(StringType).as(gf_trd_main_ent_cncl_id),
          col(gf_management_portfolio_id), lit(nullString).cast(StringType).as(g_same_ent_trd_pfl_ind_type),
          col(gf_trading_desk_id), col(g_internal_operation_ind_type), lit(nullString).cast(StringType).as(gf_trd_room_id),
          lit(nullString).cast(StringType).as(gf_trd_channel_id), lit(nullString).cast(StringType).as(gf_trd_channel_name),
          lit(nullString).cast(StringType).as(gf_exotic_trd_ind_type), lit(nullString).cast(StringType).as(gf_trd_reporting_party_id),
          lit(nullString).cast(StringType).as(gf_trd_mandty_report_ind_type), lit(nullString).cast(StringType).as(gf_block_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_block_trd_reason_typo_desc), lit(nullString).cast(StringType).as(gf_gl_hedge_objv_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_pend_to_clear_trd_ind_type), lit(nullString).cast(StringType).as(gf_off_market_price_ind_type),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_franchise_amount), lit(nullString).cast(StringType).as(gf_trd_franchise_currency_id),
          lit(nullString).cast(StringType).as(gf_franchise_trd_desk_id), lit(nullString).cast(StringType).as(gf_trd_cnfrm_platform_name),
          lit(nullString).cast(StringType).as(gf_trd_exec_platform_typo_desc), lit(nullString).cast(StringType).as(gf_trd_exec_platform_name),
          lit(nullString).cast(StringType).as(gf_trd_tmpl_typo_desc), lit(nullString).cast(StringType).as(gf_flex_trd_ind_type),
          lit(nullString).cast(TimestampType).as(gf_trd_exec_time_date), lit(nullString).cast(TimestampType).as(gf_trd_clearing_time_date),
          lit(nullString).cast(StringType).as(gf_trd_clearing_status_desc), lit(nullString).cast(TimestampType).as(gf_trd_last_modification_date),
          lit(nullString).cast(StringType).as(gf_trd_status_typo_desc), lit(nullString).cast(StringType).as(gf_trd_vldtn_status_type),
          lit(nullString).cast(StringType).as(gf_trd_vldtn_status_type_desc), lit(nullString).cast(StringType).as(gf_special_cond_trd_ind_type),
          lit(nullString).cast(DecimalType(five, number_zero)).as(gf_trd_leg_number),
          lit(nullString).cast(DecimalType(five, number_zero)).as(gf_trd_phase_number),
          lit(nullString).cast(StringType).as(gf_trd_comment_desc), lit(nullString).cast(StringType).as(gf_financial_trd_typo_desc),
          lit(nullString).cast(StringType).as(gf_trd_delta_ind_type), lit(nullString).cast(DecimalType(twenty, number_zero)).as(gf_trd_delta_number),
          lit(nullString).cast(StringType).as(gf_trd_secnd_ent_id), lit(nullString).cast(StringType).as(gf_trd_secnd_ent_cncl_id),
          lit(nullString).cast(StringType).as(gf_trd_secnd_ent_pfolio_id), lit(nullString).cast(StringType).as(gf_secnd_ent_branch_id),
          lit(nullString).cast(StringType).as(gf_ctpty_gl_hedge_ind_type), lit(nullString).cast(StringType).as(gf_ctpty_brkr_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_trd_rptg_ctprl_typo_desc), lit(nullString).cast(StringType).as(gf_ctpty_trdr_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_trd_ctpty_objective_desc), lit(nullString).cast(StringType).as(gf_trd_ctpty_strategy_id),
          lit(nullString).cast(StringType).as(gf_blind_brkr_ind_type), lit(nullString).cast(StringType).as(gf_blind_brkr_side_type),
          lit(nullString).cast(StringType).as(gf_markit_trdr_id), lit(nullString).cast(StringType).as(gf_trd_srce_sys_salesman_id),
          lit(nullString).cast(StringType).as(gf_trd_risk_section_typo_desc), lit(nullString).cast(StringType).as(gf_trd_rptg_regime_name),
          lit(nullString).cast(StringType).as(gf_trd_supervisory_body_name), lit(nullString).cast(StringType).as(gf_trd_gl_portfolio_id),
          lit(nullString).cast(StringType).as(gf_fx_fwd_speculative_ind_type), lit(nullString).cast(StringType).as(gf_main_ent_trd_objective_desc),
          lit(nullString).cast(StringType).as(gf_fx_rt_negotiation_ind_type), lit(nullString).cast(StringType).as(gf_trd_negotiation_typo_desc),
          lit(nullString).cast(DateType).as(gf_trd_last_negotiation_date), col(gf_trd_cpt_cnsptn_pr_ind_type),
          lit(nullString).cast(StringType).as(gf_prin_elem_trd_ind_type), lit(nullString).cast(StringType).as(gf_collateral_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_exercise_trd_ind_type), lit(nullString).cast(StringType).as(gf_trd_tnspcy_exmp_ind_type),
          lit(nullString).cast(StringType).as(gf_exclv_pfolio_acc_ind_type), lit(nullString).cast(StringType).as(gf_hedged_prd_typo_desc),
          lit(nullString).cast(StringType).as(gf_clearing_exmp_trd_ind_type), lit(nullString).cast(StringType).as(gf_eu_party_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_uk_party_trd_ind_type), lit(nullString).cast(StringType).as(gf_usa_party_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_trd_key_info_doc_ind_type), lit(nullString).cast(StringType).as(gf_benchmark_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_package_trd_ind_type), lit(nullString).cast(StringType).as(gf_exch_phys_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_ref_price_trd_ind_type), lit(nullString).cast(StringType).as(gf_nego_trans_lqd_ind_type),
          lit(nullString).cast(StringType).as(gf_nego_trans_ilqd_ind_type), lit(nullString).cast(StringType).as(gf_nego_trd_cond_oth_ind_type),
          lit(nullString).cast(StringType).as(gf_trd_abv_spec_size_ind_type), lit(nullString).cast(StringType).as(gf_trd_ilqd_instm_ind_type),
          lit(nullString).cast(StringType).as(gf_short_sell_trd_ind_type), lit(nullString).cast(StringType).as(gf_brk_claus_typo_desc),
          lit(nullString).cast(StringType).as(gf_brk_claus_notif_p_type), col(gf_break_clause_date_desc),
          lit(nullString).cast(StringType).as(gf_break_clause_frequency_desc), lit(nullString).cast(StringType).as(gf_break_clause_ref_info_desc),
          lit(nullString).cast(StringType).as(gf_general_colat_ind_type), lit(nullString).cast(StringType).as(g_bbva_trdr_id),
          lit(nullString).cast(StringType).as(gf_trdr_front_sys_id), lit(nullString).cast(StringType).as(gf_bilateral_ctpty_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_bilateral_ctpty_cnncl_id), lit(nullString).cast(StringType).as(gf_bilateral_ctpty_desc),
          lit(nullString).cast(StringType).as(gf_clearing_house_srce_sys_id), lit(nullString).cast(StringType).as(gf_clearing_house_cnncl_id),
          lit(nullString).cast(StringType).as(g_clearing_house_type), lit(nullString).cast(StringType).as(gf_clearing_house_desc),
          lit(nullString).cast(StringType).as(gf_clearing_broker_id), lit(nullString).cast(StringType).as(gf_block_trd_agnt_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_block_trd_agnt_cnncl_id), lit(nullString).cast(StringType).as(gf_trd_brkr_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_trd_brkr_cnncl_id), lit(nullString).cast(StringType).as(gf_trd_brkr_desc),
          lit(nullString).cast(StringType).as(gf_trans_dealer_srce_sys_id), lit(nullString).cast(StringType).as(gf_trans_dealer_cnncl_id),
          lit(nullString).cast(StringType).as(gf_trd_lqdty_prvd_srce_sys_id), lit(nullString).cast(StringType).as(gf_trd_lqdty_prvd_cnncl_id),
          lit(nullString).cast(StringType).as(gf_trd_settl_agnt_srce_sys_id), lit(nullString).cast(StringType).as(gf_trd_settl_agent_desc),
          lit(nullString).cast(StringType).as(gf_trd_settl_agnt_cnncl_id), lit(nullString).cast(StringType).as(gf_ctpty_benef_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_ctpty_benef_cnncl_id), lit(nullString).cast(StringType).as(gf_trd_calc_agnt_srce_sys_id),
          lit(nullString).cast(StringType).as(gf_trd_calc_agnt_cnncl_id), lit(nullString).cast(StringType).as(g_prod_bbg_ticker_id),
          lit(nullString).cast(StringType).as(gf_prod_exch_mkt_id), col(g_mic_id), lit(nullString).cast(StringType).as(gf_market_origin_name),
          lit(nullString).cast(StringType).as(gf_trd_trpty_agnt_srce_sys_id), lit(nullString).cast(StringType).as(gf_trd_trpty_agnt_cnncl_id),
          lit(nullString).cast(StringType).as(gf_trd_ori_event_type), col(gf_trd_ori_event_typo_desc),
          lit(nullString).cast(StringType).as(gf_regy_event_srce_sys_desc),
          lit(nullString).cast(StringType).as(g_detail_event_id),
          lit(nullString).cast(StringType).as(gf_cnfrm_sta_trd_evt_type_desc),
          lit(nullString).cast(StringType).as(gf_rlse_sta_trd_evt_type_desc),
          lit(nullString).cast(StringType).as(gf_prdef_mod_cncl_fld_ind_type),
          lit(nullString).cast(StringType).as(gf_prdef_mod_cncl_fld_cs_desc),
          lit(nullString).cast(StringType).as(gf_prdef_mod_cncl_fld_jst_desc),
          lit(nullString).cast(StringType).as(gf_trd_maturity_type),
          lit(nullString).cast(DateType).as(gf_trd_change_agree_date), col(gf_trd_change_effective_date),
          lit(nullString).cast(StringType).as(gf_trd_event_time_date), lit(nullString).cast(DateType).as(gf_trd_change_valuation_date),
          lit(nullString).cast(StringType).as(gf_new_trd_id), lit(nullString).cast(StringType).as(gf_trd_transferor_id),
          lit(nullString).cast(StringType).as(gf_trd_transferor_desc), lit(nullString).cast(StringType).as(gf_trd_transferee_id),
          lit(nullString).cast(StringType).as(gf_trd_transferee_desc), lit(nullString).cast(StringType).as(gf_remaining_ctpty_id),
          lit(nullString).cast(StringType).as(gf_event_reason_desc), lit(nullString).cast(StringType).as(gf_event_reason_id),
          lit(nullString).cast(StringType).as(gf_event_vldtn_last_user_id), lit(nullString).cast(StringType).as(gf_cancelled_trade_id),
          lit(nullString).cast(StringType).as(gf_user_cancel_id), lit(nullString).cast(StringType).as(gf_event_cnncl_ind_type),
          lit(nullString).cast(StringType).as(gf_evt_cnncl_reevent_ind_type), lit(nullString).cast(StringType).as(gf_event_vldtn_status_desc),
          lit(nullString).cast(StringType).as(gf_trd_seller_comment_desc), lit(nullString).cast(StringType).as(gf_trd_buyer_comment_desc),
          lit(nullString).cast(StringType).as(g_prior_trd_usi_id), lit(nullString).cast(StringType).as(gf_related_trd_id),
          lit(nullString).cast(StringType).as(gf_cstm_trd_ind_type), lit(nullString).cast(StringType).as(gf_cstm_trd_object_desc),
          lit(nullString).cast(StringType).as(gf_cstm_trd_value_desc), lit(nullString).cast(StringType).as(gf_cstm_trd_cshflw_bkwd_desc),
          lit(nullString).cast(StringType).as(gf_accounting_section_triad_id), lit(nullString).cast(StringType).as(gf_autocapture_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_auto_autocapture_ind_type), col(g_spot_forward_type),
          lit(nullString).cast(TimestampType).as(gf_srce_sys_audit_date), lit(nullString).cast(StringType).as(gf_srce_sys_audit_group_type),
          lit(nullString).cast(StringType).as(gf_srce_sys_audit_event_type), lit(nullString).cast(StringType).as(gf_srce_sys_audit_sbtyp_desc),
          lit(nullString).cast(StringType).as(gf_srce_sys_audit_user_id), lit(nullString).cast(DateType).as(gf_trading_desk_proc_date),
          lit(nullString).cast(StringType).as(gf_structured_prd_typo_desc), lit(nullString).cast(StringType).as(gf_structured_prd_id),
          lit(nullString).cast(StringType).as(gf_ope_modality_type), lit(nullString).cast(StringType).as(gf_trd_modality_type_desc),
          col(gf_fin_trade_price_amount).as(gf_fin_trade_price_amount), lit(nullString).cast(StringType).as(gf_trd_ctrct_price_typo_desc),
          lit(nullString).cast(DateType).as(gf_trd_spot_date), lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_spot_prc_amount),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_face_amount), lit(nullString).cast(StringType).as(gf_straddle_strat_ind_type),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_lot_size_number),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_ctrct_nominal_amount),
          col(gf_interest_rate_change_date).cast(DateType), lit(nullString).cast(DecimalType(twelve, nine)).as(gf_total_amortization_fee_per),
          col(g_renewal_type), lit(nullString).cast(StringType).as(gf_trd_fwd_strt_pyof_typo_desc), col(gf_deriv_setl_mkt_ind_type),
          col(gf_clear_cust_op_ind_type), col(gf_derv_net_dpst_ind_type), lit(nullString).cast(StringType).as(gf_distribution_channel_type),
          lit(nullString).cast(StringType).as(g_mother_trade_ind_type), lit(nullString).cast(StringType).as(gf_mother_trd_allocation_desc),
          lit(nullString).cast(StringType).as(gf_novation_mother_desc), lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_markup_amount),
          lit(nullString).cast(StringType).as(gf_interest_group_type_desc), lit(nullString).cast(StringType).as(gf_interest_grp_user_ind_type),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_fee_amount), lit(nullString).cast(StringType).as(g_fee_curncy_id),
          lit(nullString).cast(DateType).as(gf_trd_fee_date), lit(nullString).cast(StringType).as(gf_modified_operation_star_id),
          lit(nullString).cast(StringType).as(gf_trade_authorization_type_id), lit(nullString).cast(StringType).as(gf_trade_authorization_user_id),
          lit(nullString).cast(TimestampType).as(gf_trade_authorization_date), lit(nullString).cast(StringType).as(gf_trade_authorization_desc),
          lit(nullString).cast(StringType).as(gf_trade_auth_observ_desc), lit(nullString).cast(StringType).as(gf_auth_modl_clsfn_type_desc),
          lit(nullString).cast(StringType).as(gf_class_modl_auth_error_id), lit(nullString).cast(StringType).as(gf_comm_actvy_link_ctrct_id),
          lit(nullString).cast(StringType).as(gf_trcblty_info_desc), lit(nullString).cast(StringType).as(gf_mifid_trader_id),
          lit(nullString).cast(StringType).as(gf_strategy_main_trade_id), lit(nullString).cast(StringType).as(gf_oper_exercise_execution_id),
          lit(nullString).cast(StringType).as(gf_cb_tnspcy_exmp_tran_id), lit(nullString).cast(StringType).as(gf_mifid_manager_user_id),
          lit(nullString).cast(StringType).as(g_mifid_isin_id), lit(nullString).cast(StringType).as(gf_mifid_trd_exec_pltfm_name),
          lit(nullString).cast(StringType).as(gf_algo_decision_id), lit(nullString).cast(StringType).as(gf_otc_post_trade_flag_id),
          lit(nullString).cast(StringType).as(gf_pre_trade_waiver_flag_id), lit(nullString).cast(StringType).as(g_analytics_currency_id),
          lit(nullString).cast(StringType).as(gf_gl_account_mark_type), lit(nullString).cast(StringType).as(gf_cls_participation_ind_type),
          lit(nullString).cast(StringType).as(gf_loan_or_gntee_ind_type), lit(nullString).cast(DateType).as(gf_recouponing_date),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_operation_principal_amount),
          lit(nullString).cast(StringType).as(g_trd_rstru_replaced_ind_type),
          lit(nullString).cast(StringType).as(gf_trd_source_trading_sys_id), lit(nullString).cast(StringType).as(gf_user_group_id),
          lit(nullString).cast(StringType).as(gf_source_obsv_desc), lit(nullString).cast(StringType).as(gf_dest_obsv_desc),
          lit(nullString).cast(TimestampType).as(gf_trd_capture_time_date), lit(nullString).cast(StringType).as(gf_execution_algorithm_desc),
          lit(nullString).cast(StringType).as(g_trd_allocate_ind_type), lit(nullString).cast(StringType).as(gf_offer_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_funding_cost_amount),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_loan_fund_cost_amount),
          lit(nullString).cast(StringType).as(gf_trd_linked_trd_id), lit(nullString).cast(StringType).as(gf_trd_risk_proposal_id),
          lit(nullString).cast(StringType).as(gf_trd_synergy_branch_id), lit(nullString).cast(DecimalType(twelve, eight)).as(gf_brokerage_fee_amount),
          lit(nullString).cast(StringType).as(gf_brokerage_fee_id), lit(nullString).cast(TimestampType).as(gf_trd_trading_time_date),
          lit(nullString).cast(StringType).as(gf_trade_hms_date), lit(nullString).cast(StringType).as(gf_geography_id),
          lit(nullString).cast(StringType).as(gf_distribution_centre_desc), lit(nullString).cast(StringType).as(gf_trade_franchise_table_desc),
          lit(nullString).cast(StringType).as(gf_trd_frnt_sys_ent_label_name), lit(nullString).cast(StringType).as(gf_event_srce_sys_desc),
          lit(nullString).cast(StringType).as(gf_event_comment_desc), lit(nullString).cast(DateType).as(gf_mifid_date),
          lit(nullString).cast(StringType).as(gf_mifid_time_date), lit(nullString).cast(DecimalType(twelve, six)).as(gf_mifid_time_ms_number),
          lit(nullString).cast(StringType).as(gf_mifid_instrument_id), lit(nullString).cast(StringType).as(gf_rfpt_waive_pt_tran_ind_type),
          lit(nullString).cast(StringType).as(gf_oilq_waive_pt_tran_ind_type), lit(nullString).cast(StringType).as(gf_size_waive_pt_tran_ind_type),
          lit(nullString).cast(StringType).as(gf_nliq_waive_pt_tran_ind_type), lit(nullString).cast(StringType).as(gf_pric_waive_pt_tran_ind_type),
          lit(nullString).cast(StringType).as(gf_iliq_waive_pt_tran_ind_type), lit(nullString).cast(StringType).as(gf_mifid_sell_type),
          lit(nullString).cast(StringType).as(gf_benc_otc_pst_trans_ind_type), lit(nullString).cast(StringType).as(gf_tpac_otc_pst_trans_ind_type),
          lit(nullString).cast(StringType).as(gf_xfph_otc_pst_trans_ind_type), lit(nullString).cast(StringType).as(g_mifid_dpo_rp_prntrd_ind_type),
          lit(nullString).cast(StringType).as(g_fx_fwd_comm_hedge_ind_type), lit(nullString).cast(StringType).as(gf_decision_algorithm_name),
          lit(nullString).cast(StringType).as(gf_execution_algorithm_name), lit(nullString).cast(StringType).as(g_mifid_ecb_exec_ind_type),
          lit(nullString).cast(StringType).as(gf_trd_native_event_id), lit(nullString).cast(TimestampType).as(gf_trd_native_event_time_date),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_costs_charges_amount),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trd_dist_franchise_amount),
          lit(nullString).cast(StringType).as(gf_dtcc_event_desc), lit(nullString).cast(StringType).as(gf_last_market_operation_type),
          lit(nullString).cast(StringType).as(gf_last_market_operation_desc), lit(nullString).cast(StringType).as(gf_market_operation_type_id),
          lit(nullString).cast(DateType).as(gf_market_operation_first_date), lit(nullString).cast(StringType).as(gf_market_operation_type),
          lit(nullString).cast(StringType).as(gf_broker_sec_id), lit(nullString).cast(StringType).as(gf_advise_ind_type),
          lit(nullString).cast(DateType).as(gf_trd_cont_event_actual_date), lit(nullString).cast(DateType).as(gf_trd_cont_event_ses_lgn_date),
          lit(nullString).cast(DateType).as(gf_trd_cont_event_date), lit(nullString).cast(StringType).as(gf_trd_cont_event_action_type),
          lit(nullString).cast(StringType).as(gf_market_index_id), lit(nullString).cast(StringType).as(gf_instm_issuance_reference_id),
          lit(nullString).cast(StringType).as(gf_trd_orig_ft_app_id),
          lit(nullString).cast(StringType).as(gf_ft_orig_app_trd_id),
          lit(nullString).cast(StringType).as(gf_trd_orig_lst_app_id),
          lit(nullString).cast(StringType).as(gf_lst_orig_app_trd_id),
          lit(nullString).cast(StringType).as(g_trd_master_contract_ind_type),
          lit(nullString).cast(StringType).as(g_last_change_user_id),
          lit(nullString).cast(TimestampType).as(gf_trade_repo_data_insert_date),
          lit(nullString).cast(StringType).as(gf_min_notice_period_id),
          lit(nullString).cast(DateType).as(gf_closest_callback_date),
          lit(nullString).cast(StringType).as(g_central_sec_dpsty_bic_id),
          lit(nullString).cast(StringType).as(gf_lender_agent_id),
          lit(nullString).cast(StringType).as(g_country_resp_gen_uti_id),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_fixed_rebate_rate_per),
          lit(nullString).cast(StringType).as(gf_flt_rebate_rate_index_type),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_flt_rebate_rate_spread_per),
          lit(nullString).cast(DecimalType(nine, number_zero)).as(gf_murex_operation_id),
          lit(nullString).cast(StringType).as(gf_abaco_trade_id),
          lit(nullString).cast(StringType).as(gf_invty_rmov_mx_tdrg_ind_type),
          lit(nullString).cast(StringType).as(gf_npc_assigned_product_id),
          lit(nullString).cast(StringType).as(gf_star_eqv_product_group_id),
          lit(nullString).cast(StringType).as(gf_star_equivalent_product_id),
          lit(nullString).cast(StringType).as(gf_trade_bbva_mgr_invs_fund_id),
          lit(nullString).cast(StringType).as(gf_year_id),
          lit(nullString).cast(DecimalType(six, number_zero)).as(gf_month_number),
          lit(nullString).cast(DecimalType(six, number_zero)).as(gf_day_number),
          lit(nullString).cast(DateType).as(gf_break_clause_date),
          lit(nullString).cast(DateType).as(gf_break_clause_notif_date),
          col(g_triparty_trade_ind_type),
          lit(nullString).cast(StringType).as(g_qis_trade_ind_type),
          lit(nullString).cast(StringType).as(gf_fx_swap_far_leg_trade_id),
          lit(nullString).cast(StringType).as(gf_pse_trade_id),
          lit(nullString).cast(StringType).as(g_dvsp_trade_settl_ind_type),
          lit(nullString).cast(StringType).as(gf_trade_seniority_desc),
          lit(nullString).cast(DateType).as(gf_break_clause_next_date),
          lit(nullString).cast(StringType).as(gf_break_clause_next_typ_desc),
          col(g_gl_classification_type),
          lit(nullString).cast(StringType).as(g_market_type),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trade_payment_amount),
          lit(nullString).cast(StringType).as(g_trade_pymt_currency_id),
          lit(nullString).cast(DecimalType(twenty_six, six)).as(gf_trade_receipt_amount),
          lit(nullString).cast(StringType).as(g_trade_rcpt_currency_id),
          lit(nullString).cast(StringType).as(g_cancelb_option_trd_ind_type),
          lit(nullString).cast(StringType).as(gf_trd_cncb_opt_parties_desc),
          lit(nullString).cast(DateType).as(gf_trd_cancelled_next_date),
          lit(nullString).cast(StringType).as(gf_trd_cancelled_term_type),
          lit(nullString).cast(DecimalType(twenty_six, eight)).as(gf_trd_prdt_signed_qty_number),
          lit(nullString).cast(DecimalType(twenty_six, eight)).as(gf_fin_trade_clos_price_amount),
          lit(nullString).cast(StringType).as(gf_clearing_op_derv_settl_type),
          lit(odate.replace("-", emptyString)).as(gf_odate_date_id),
          current_timestamp.as(gf_audit_date), concat(lit(odate.replace("-", emptyString)),
            lit(Random.alphanumeric take 11 mkString)).as(gf_run_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
}
```

## Homologate

```scala
import Constants._
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.{DataFrame, SparkSession, Column}
import org.apache.spark.sql.functions.{col, when, substring, concat, lit, lpad, upper}

class Homologate(odate: String) {

  def fieldsHomologate(df: DataFrame, valuesToReplace: List[String], colsToIgnore: List[String]): DataFrame = {
    try {
      df.select(df.columns.map(x => if (!colsToIgnore.contains(x)) {
        when(col(x).isin(valuesToReplace: _*), nullString)
          .otherwise(col(x)).as(x)
      } else {
        col(x)
      }): _*)
    } catch {
      case e: Exception => balance_utility + ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def contractHomologateBank(df: DataFrame, colsToHomologate: List[String]): DataFrame = {
    try {
      df.select(df.columns.map(x => if (colsToHomologate.contains(x)) {
        when(substring(col(x), one, four).isin(entity,entity0007,entity0005,entity9999),(concat(lit(country),col(x))))
          .otherwise(concat(lit(country_entity),col(x))).as(x)
      }
      else {
        col(x)
      }): _*)
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
  
  def clientHomologateBank(df: DataFrame, colsToHomologate: List[String]): DataFrame = {
    try {
      df.select(df.columns.map(x => if (colsToHomologate.contains(x)) {
        when(substring(col(x), one, six).isin(MX0005, MX0007, MX9999), col(x))
          .otherwise(concat(lit(country_entity), lpad(col(x), nine, zero_string)))
          .as(x)
      }
      else {
        col(x)
      }): _*)
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }
}
```
